## 1. What is Python?

**Analogy:** Imagine you want to automate scanning 10,000 log lines every morning to flag suspicious IPs. You could do it by hand — or you could write a short Python script that does it in seconds while you drink your coffee. Python is a language designed to let humans write instructions that computers can execute, in words that humans can actually read.

Python is a **high-level, interpreted programming language**. Let's break that down:

- **High-level** means Python looks like English. You write `print("Scan complete")` instead of raw binary instructions. The computer handles the translation.
- **Interpreted** means Python runs your code line by line, immediately, without a compilation step. You write a line, run it, see the result. This is ideal for security work where you often need to test a small idea quickly during an incident.

### Why Security Professionals Use Python

| Reason | What it means in practice |
|---|---|
| **Readable syntax** | You can audit someone else's script quickly and spot malicious logic |
| **Batteries included** | The standard library has built-in tools for networking (`socket`), encryption (`hashlib`, `ssl`), file handling, and log parsing — no install needed |
| **Interpreted** | Test a one-liner during incident response; no compile-run-debug cycle |
| **Huge ecosystem** | Thousands of security-specific packages: `scapy`, `requests`, `paramiko`, `pwntools` |
| **Industry standard** | Every SOC, pentest firm, and cloud team uses Python for automation |

### Python vs other options

You might wonder: why not use Bash, PowerShell, or another language?

| Language | Good at | Weak at |
|---|---|---|
| **Python** | Automation, parsing, scripting, prototyping | Low-level memory control |
| **Bash** | Quick one-liners, file system operations | Readable logic, data structures |
| **PowerShell** | Windows administration | Cross-platform work |
| **C/C++** | Speed, system-level exploits | Rapid development, readability |

You will learn Bash and PowerShell in the Operating Systems module. Python is your primary scripting language for the whole programme.

## 2. Setting Up Your Environment

**Analogy:** Before you can cook, you need a kitchen. Before you can write Python, you need the Python interpreter installed on your machine. The interpreter is the engine that reads your code and executes it.

We recommend **Python 3.10 or higher**. Avoid Python 2 — it reached end-of-life in 2020 and is no longer maintained.

### Windows

**Option A — Installer:**
1. Go to [python.org/downloads](https://www.python.org/downloads/windows/)
2. Download the latest Python 3.x installer
3. Run the installer — **tick the box "Add Python to PATH"** before clicking Install

**Option B — Command line (if you have winget):**
```powershell
winget install Python.Python.3.12
```

### macOS

**Option A — Installer:**
1. Go to [python.org/downloads](https://www.python.org/downloads/macos/)
2. Download the macOS universal2 installer

**Option B — Homebrew (recommended if you use the terminal regularly):**
```bash
brew install python
```

> **Note for macOS users:** macOS comes with Python 2 pre-installed. Always type `python3` (not `python`) in your terminal to make sure you are using the right version.

### Linux (Ubuntu / Debian)

Most Linux distributions include Python 3 by default. To ensure you also have `pip` (the package manager):
```bash
sudo apt update
sudo apt install python3 python3-pip python3-venv
```

### Verifying Your Installation

Open your terminal and type:
```bash
python --version
```

Or on macOS/Linux:
```bash
python3 --version
```

You should see something like `Python 3.12.0`. If you get `command not found`, the installation did not complete — go back and check the PATH setting.


````python
# You can also check your Python version from INSIDE Python (useful in scripts)
import sys

major = sys.version_info.major   # should be 3
minor = sys.version_info.minor   # should be 10 or higher

print(f"Python {major}.{minor}")

if major == 3 and minor >= 10:
    print("Version check: PASS - ready for this course")
else:
    print("Version check: FAIL - please upgrade to Python 3.10 or higher")
````

## 4. pip and Package Management

**Analogy:** The Python standard library is like the tools that come with your laptop out of the box. `pip` is the app store — it lets you install thousands of additional tools that other developers have built and shared publicly on the **Python Package Index (PyPI)** at [pypi.org](https://pypi.org).

When you run `pip install requests`, it connects to PyPI, downloads the `requests` package and all its dependencies, and installs them so you can `import` them in your code.

### Core pip commands

All of these are run in your **terminal**, not inside a notebook cell:

| Command | What it does |
|---|---|
| `pip install requests` | Install a package |
| `pip install requests==2.31.0` | Install a specific version |
| `pip uninstall requests` | Remove a package |
| `pip list` | Show all installed packages and their versions |
| `pip show requests` | Show details about one package (version, author, location) |
| `pip freeze > requirements.txt` | Export all installed packages to a file |
| `pip install -r requirements.txt` | Install all packages from a requirements file |

### requirements.txt — the standard way to share dependencies

When you share a project with a colleague, you do not share the installed packages — you share a `requirements.txt` file that lists what is needed. They run `pip install -r requirements.txt` to recreate your exact environment.

Example `requirements.txt`:
```
requests==2.31.0
scapy==2.5.0
python-nmap==0.7.1
```

> **Security warning:** Never install a package from PyPI without checking it first. Typosquatting attacks (e.g., `requets` instead of `requests`) are real — attackers publish packages with names close to popular ones to steal credentials. Always verify the package name, author, and download count before installing.

### Execution order matters

Jupyter remembers everything you have run in the current session. If you define a variable in cell 5 and use it in cell 3, it works — **as long as you ran cell 5 first**. This can cause confusing bugs if you run cells out of order.

**Best practice:** Run cells from top to bottom. When in doubt, use **Kernel > Restart & Run All** to run the entire notebook fresh from the beginning.

### The execution counter

The number in `[3]:` to the left of a code cell tells you in what order cells were run. `[*]:` means the cell is currently running. This helps you track what state your notebook is in.

## 6. Your First Python Lines

**Analogy:** Before an analyst writes an incident report, they write bullet notes. Before writing a full port scanner, you write small experiments. This section runs a few lines of real Python so you leave this notebook having actually done something — not just read about it.

We will keep this brief. The next notebook (`00_Python_Syntax.ipynb`) covers the language rules in depth. Here, we just confirm that Python works and start building confidence.

### print() — your most used debugging tool

`print()` outputs text or a value to the screen. You will use it constantly: to see what a variable contains, to confirm a script reached a certain point, and to display results.

```python
print("message")      # print a text string
print(42)             # print a number
print(variable_name)  # print the value stored in a variable
```


````python
# print() - the most used function in Python

print("Hello from Python")          # text must be in quotes
print(443)                          # numbers do not need quotes
print("Open ports:", [22, 80, 443]) # you can print multiple things separated by commas
````

````python
# Variables store values so you can use them later
# Python figures out the type automatically - you do not declare it

target_ip        = "10.0.0.55"   # str  - text in quotes
scan_duration_ms = 342           # int  - whole number
cvss_score       = 9.8           # float - decimal number
firewall_enabled = True          # bool - True or False

print(target_ip)
print(scan_duration_ms)
print(cvss_score)
print(firewall_enabled)
````

````python
# Python can do arithmetic
# You will use this for: port calculations, CVSS scoring, log analysis, timestamps

total_ports      = 65535
scanned_ports    = 1000
remaining_ports  = total_ports - scanned_ports
scan_percentage  = (scanned_ports / total_ports) * 100

print(f"Scanned {scanned_ports} of {total_ports} ports")
print(f"Coverage: {scan_percentage:.1f}%")   # :.1f means 'round to 1 decimal place'
print(f"Remaining: {remaining_ports} ports")
````

````python
# A list stores multiple values in order
# You will learn lists in depth in the Collections module
# For now, just see that Python can hold a collection of ports in one variable

open_ports = [22, 80, 443, 3389, 8080]  # list of integers

print("Open ports found:", open_ports)
print("Number of open ports:", len(open_ports))  # len() counts items
print("First port in list:", open_ports[0])       # [0] accesses the first item
````
