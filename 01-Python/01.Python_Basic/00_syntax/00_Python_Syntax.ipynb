## 1. Design Philosophy — Why Python Looks the Way It Does

**Analogy:** Some programming languages are like writing a legal contract — dense, full of
mandatory symbols, and hard to read without training. Python was deliberately designed to look
more like plain English. The goal was that a colleague who does not know Python should be able
to read your script and roughly understand what it does.

This is summarised in a document called **The Zen of Python** (PEP 20), written by Tim Peters.
The most relevant principles for security work:

```
Beautiful is better than ugly.
Explicit is better than implicit.
Simple is better than complex.
Readability counts.
There should be one — and preferably only one — obvious way to do it.
```

**Why this matters in cybersecurity:** Security teams read each other's scripts constantly —
for code review, incident response, and auditing. Unreadable code is a security risk in itself.
A confusing script can hide malicious logic or make a defensive tool impossible to maintain.

### Python vs other languages

Here is the same logic written in Java and Python:

**Java:**
```java
if (port == 80) {
    System.out.println("HTTP port open");
}
```

**Python:**
```python
if port == 80:
    print("HTTP port open")
```

Python removes curly braces `{}`, semicolons `;`, and type declarations.
Instead, it uses a colon `:` to open a block and **indentation** to define what is inside it.

## 2. Indentation — How Python Structures Code Blocks

**Analogy:** Think of a military briefing document. Sections are at the top level.
Sub-tasks are indented under their section. Sub-sub-tasks are indented further.
The visual structure tells you at a glance what belongs where. Python works the same way —
the indentation level tells Python what code belongs inside which block.

In Python, indentation is **not optional** — it is part of the syntax.
Getting it wrong causes an `IndentationError` and your code will not run.

**The rule:** After any line ending with `:`, all code belonging to that block must be
indented by the same amount. The standard is **4 spaces per level**.

> **Important:** Use spaces, not tabs. Most code editors (VS Code, PyCharm) convert Tab
> key presses to 4 spaces automatically. Mixing tabs and spaces causes invisible bugs.


````python
# One level of indentation: code inside the if block is indented 4 spaces
port = 22

if port == 22:
    print("Port 22 is open")   # inside the if block — 4 spaces
    print("Service: SSH")      # inside the if block — 4 spaces

print("Scan complete")         # outside the if block — 0 spaces (always runs)
````

````python
# Two levels of indentation: for loop contains an if block
# Don't worry about the 'for' syntax yet - focus only on the structure

open_ports = [22, 80, 443]

for port in open_ports:           # Level 1 — starts a for block
    if port == 22:                # Level 2 — inside the for block
        print(f"{port}: SSH")     # Level 3 — inside the if AND the for block
    elif port == 80:
        print(f"{port}: HTTP")    # Level 3
    else:
        print(f"{port}: HTTPS")   # Level 3

print("All ports scanned")        # Level 1 — outside the for block entirely
````

### Warning - Common Indentation Errors

These errors are the most common bugs beginners hit. Read them carefully.

**Error 1 — Missing indentation:** Code after `:` is at the same level as the colon line.

**Error 2 — Inconsistent indentation:** Lines inside the same block have different
numbers of spaces (e.g. one line has 4 spaces, the next has 2).

Both produce `IndentationError: expected an indented block`.


````python
# WRONG — missing indentation (commented out so this cell runs)
# if port == 22:
# print("SSH")    # should be indented 4 spaces

# CORRECT
port = 22
if port == 22:
    print("SSH")   # 4 spaces — belongs inside the if block
````

````python
# WRONG — inconsistent indentation (commented out)
# if port == 22:
#      print("SSH")    # 5 spaces
#    print("Secure")   # 3 spaces — IndentationError!

# CORRECT — all lines in the same block use exactly 4 spaces
port = 22
if port == 22:
    print("SSH")     # 4 spaces
    print("Secure")  # 4 spaces — consistent
````

## 3. Comments — Documenting Your Code

**Analogy:** An analyst who writes a detailed incident report is more valuable than one who
scribbles illegible notes. Comments are your incident report for future readers — including
yourself, six months from now, trying to remember why you wrote something a certain way.

A **comment** is text in your code that Python **completely ignores**. It exists only for
the programmer reading the code.

> "Code tells you **how**. Comments tell you **why**." — Jeff Atwood

### Why Comments Matter in Cybersecurity

Security scripts are read during audits, peer review, and incident response. A port scanner
without comments is just a black box. A port scanner **with** comments explains:
- Why a specific port range was chosen
- Why a particular IP is hardcoded
- What authorisation this script operates under
- Why a security check is implemented a certain way

### Type 1: Single-line comment (`#`)

The `#` character starts a comment. Everything from `#` to the end of the line is ignored.


````python
# This entire line is a comment — Python ignores it completely
print("Connection attempt")  # inline comment — only from # onwards is ignored

# Security context: always document why a port matters
ssh_port   = 22    # SSH — encrypted remote access, acceptable to expose
telnet_port = 23   # Telnet — unencrypted, must be blocked at the firewall
rdp_port   = 3389  # RDP — high-value target, restrict to VPN only
````

### Type 2: Multi-line comment (triple quotes)

For longer explanations — such as a script header or a complex explanation — use triple
quotes `"""`. Python reads this as a string but does not execute it unless it is assigned
to a variable.

The most common use is at the **top of a script** to document what it does, who wrote it,
and under what authorisation it operates.


````python
"""
Port Scanner Script
Author  : j.smith
Date    : 2024-01-15
Version : 1.0

Purpose:
  Scans common service ports on a target host and identifies open services.

Authorisation:
  Authorised for internal use on the 192.168.0.0/16 network only.
  Do NOT run against external IPs without written authorisation.

Ports scanned: 22 (SSH), 80 (HTTP), 443 (HTTPS), 3389 (RDP)
"""

print("Script started")   # this line actually runs
````

### What makes a good comment?

**Bad comment — just repeats the code:**
```python
port = 80  # set port to 80
```

**Good comment — explains why:**
```python
port = 80  # HTTP default port — unencrypted, flagged for review
```

**Good comment — warns about a constraint:**
```python
# WARNING: do not scan external IPs without written authorisation
target_ip = "192.168.1.1"  # internal range only — authorised
```

**Good comment — explains a non-obvious technical decision:**
```python
# Using a set instead of a list for O(1) lookup
# Matters when checking 1M IPs against the blocklist
blocked_ips = {"203.0.113.5", "198.51.100.99"}
```

## 4. Quotation — Writing Text in Python

**Analogy:** In a written report, you use quotation marks to show where a quoted phrase
begins and ends. Python works the same way — quotes tell Python: "this is text data, not
code to execute".

Any piece of text in Python is called a **string**. Strings must be wrapped in quotes.

### Three types of quotes

| Type | Syntax | Best used for |
|---|---|---|
| Single quotes | `'text'` | Short text with no apostrophes |
| Double quotes | `"text"` | Short text — the preferred standard |
| Triple quotes | `"""text"""` or `'''text'''` | Multi-line text, docstrings |

**The golden rule:** open and close with the **same** type of quote.


````python
# All three quote styles produce valid strings
username   = 'j.smith'                    # single quotes
target_ip  = "10.0.0.55"                  # double quotes — preferred
scan_notes = """Target: 10.0.0.55
Status: Open
Risk: High"""                              # triple quotes — preserves line breaks

print(username)
print(target_ip)
print(scan_notes)
````

### Choosing the right quote type

Single and double quotes are interchangeable — but the right choice avoids needing
an escape character.

**Rule:** use the quote type that does **not** appear in your text.


````python
# Text contains an apostrophe (single quote inside)
warning_a = "Don't scan this IP without authorisation"  # double quotes — clean
warning_b = 'Don\'t scan this IP without authorisation' # single quotes — needs escape

# Text contains double quotes
label_a = 'The protocol is called "HTTPS"'   # single quotes — clean
label_b = "The protocol is called \"HTTPS\""  # double quotes — needs escape

print(warning_a)
print(warning_b)   # same output — different way to write it
print(label_a)
print(label_b)     # same output — different way to write it
````

### Escape characters

The backslash `\` lets you write special characters inside a string:

| Escape | Meaning | Example output |
|---|---|---|
| `\n` | New line | Moves to the next line |
| `\t` | Tab | Adds horizontal spacing |
| `\'` | Single quote | Apostrophe inside single-quoted string |
| `\"` | Double quote | Double quote inside double-quoted string |
| `\\` | Backslash | A literal `\` character |


````python
# Escape characters in a formatted scan report
print("=== Scan Report ===")
print("Host:\t192.168.1.100")    # \t adds a tab for alignment
print("Port:\t22")
print("Status:\tOpen")
print()
print("Line 1\nLine 2\nLine 3")    # \n inserts a line break
print()
print("Path: C:\\Users\\analyst")  # \\ to write a literal backslash (Windows paths)
````

````python
# Triple quotes preserve formatting — no \n needed
scan_report = """
=== SECURITY SCAN REPORT ===
Target IP  : 192.168.1.100
Analyst    : j.smith
Date       : 2024-01-15

Open Ports :
  22   SSH    (acceptable)
  80   HTTP   (review needed)
  443  HTTPS  (acceptable)

Risk Level : LOW
============================
"""
print(scan_report)
````

## 5. Naming Conventions (PEP 8)

**Analogy:** In a SOC, every incident gets a standardised ticket number, every log file
follows a naming format, every analyst follows the same report template. Consistency makes
the whole team faster. Python's PEP 8 style guide serves the same purpose — it standardises
how code is named so any Python developer can read any other developer's code.

**PEP 8** (Python Enhancement Proposal 8) is the official Python style guide. Following
it is not mandatory — your code will run either way — but it is **expected** in professional
and team environments.

### Naming rules by type

| What you are naming | Convention | Example |
|---|---|---|
| Variable | `snake_case` | `target_ip`, `failed_attempts` |
| Function | `snake_case` | `scan_port()`, `parse_log()` |
| Constant | `UPPER_SNAKE_CASE` | `MAX_RETRIES`, `DEFAULT_PORT` |
| Class | `PascalCase` | `FirewallRule`, `ScanResult` |

### Rules for valid names

| Allowed | Not allowed |
|---|---|
| Letters `a-z`, `A-Z` | Spaces: `target ip` |
| Digits `0-9` (not first) | Hyphens: `target-ip` |
| Underscore `_` | Starting with a digit: `1port` |
| | Special characters: `@`, `!`, `#`, `$` |
| | Python reserved words: `if`, `for`, `import` |


````python
# WRONG — valid Python but violates PEP 8
TargetIP        = "192.168.1.1"   # looks like a class, not a variable
FAILEDATTEMPTS  = 3               # looks like a constant, not a variable
p               = 22              # meaningless single letter

# CORRECT — PEP 8 compliant
target_ip        = "192.168.1.1"   # snake_case variable — clear and readable
failed_attempts  = 3               # snake_case variable — descriptive
port             = 22              # snake_case variable — concise but clear

# Constants: values that never change in the programme
MAX_LOGIN_ATTEMPTS = 3             # UPPER_SNAKE_CASE constant
DEFAULT_SSH_PORT   = 22            # UPPER_SNAKE_CASE constant

print(f"Target: {target_ip}:{port}")
print(f"Max login attempts: {MAX_LOGIN_ATTEMPTS}")
````

````python
# Valid names vs invalid names
port         = 80    # valid
port_number  = 80    # valid — more descriptive
port1        = 80    # valid — number allowed after first character
_private     = True  # valid — leading underscore means 'internal use'

# These would cause SyntaxError (commented out to prevent crash):
# 1port    = 80     # starts with a number
# target-ip = "..." # hyphen not allowed in names
# my ip    = "..."  # space not allowed in names

print(port, port_number, port1)
````

## 6. Reserved Words — What Python Has Already Claimed

**Analogy:** In any organisation, certain job titles are reserved — you cannot call yourself
"The CEO" if that role already exists. Python has 35 words it has already claimed a meaning
for. You cannot use them as variable names, function names, or any other identifier.

If you try to use a reserved word as a variable name, Python raises a `SyntaxError`.


````python
# See all Python reserved words programmatically
import keyword

print(f"Python has {len(keyword.kwlist)} reserved words:\n")

# Print them in a readable 5-column layout
for i, kw in enumerate(keyword.kwlist):
    print(f"{kw:<12}", end="")
    if (i + 1) % 5 == 0:
        print()
````

### Keywords grouped by purpose

You will learn what each one does throughout this course. For now, just recognise them.

**Control flow** (decisions and loops)

| Keyword | Purpose |
|---|---|
| `if`, `elif`, `else` | Conditional logic |
| `for`, `while` | Loops |
| `break`, `continue`, `pass` | Loop control |

**Functions and classes**

| Keyword | Purpose |
|---|---|
| `def` | Define a function |
| `return` | Return a value from a function |
| `class` | Define a class |
| `lambda` | Short anonymous function |

**Logic and values**

| Keyword | Purpose |
|---|---|
| `and`, `or`, `not` | Logical operators |
| `True`, `False`, `None` | Built-in constant values |
| `in`, `is` | Membership and identity tests |

**Module and scope**

| Keyword | Purpose |
|---|---|
| `import`, `from`, `as` | Loading modules (Section 9) |
| `global`, `nonlocal` | Scope control |

**Error handling** (covered in Python Advanced)

| Keyword | Purpose |
|---|---|
| `try`, `except`, `finally`, `raise`, `with` | Exception handling |


````python
# Reserved words cannot be used as variable names
# These would all cause SyntaxError — shown as comments to avoid crashing:

# for    = 80          # SyntaxError: 'for' is a reserved word
# import = "file"      # SyntaxError: 'import' is a reserved word
# class  = "firewall"  # SyntaxError: 'class' is a reserved word

# Workaround: add a descriptive suffix to make the name valid
scan_for_port  = 80           # 'scan_for_port' is valid
import_path    = "/var/log"   # 'import_path' is valid
firewall_class = "enterprise" # 'firewall_class' is valid

print(scan_for_port, import_path, firewall_class)
````

## 7. Line Continuation — Splitting Long Lines

**Analogy:** A good analyst does not write a 200-word sentence in a report. They break it
into short, readable sentences. PEP 8 recommends keeping Python lines under **79 characters**.
Long lines are hard to read, especially when reviewing security code side-by-side in a diff tool.

There are two ways to split a long line in Python.

### Method 1: Explicit continuation with `\`

Place `\` at the very end of a line (no space after it) to tell Python the line continues
on the next line.


````python
# Without continuation — one very long string (hard to read in a code review)
log_entry = "DENY | src=203.0.113.5 | dst=10.0.0.1 | port=22 | proto=TCP | rule=BLOCK-SSH-EXTERNAL"

# With continuation — split into readable chunks using \
log_entry = "DENY | src=203.0.113.5 | " \
            "dst=10.0.0.1 | port=22 | " \
            "proto=TCP | rule=BLOCK-SSH-EXTERNAL"

print(log_entry)   # same output — just more readable code
````

### Method 2: Implicit continuation inside brackets (preferred)

Inside `()`, `[]`, or `{}`, Python knows the expression continues until the closing bracket.
**This is the preferred method** — no `\` needed, and you can add comments per line.


````python
# Long list — split across lines inside []
dangerous_ports = [
    21,    # FTP — unencrypted file transfer
    23,    # Telnet — unencrypted remote access
    135,   # Windows RPC — common attack vector
    445,   # SMB — frequently exploited (EternalBlue, WannaCry)
    3389,  # RDP — remote desktop, high-value target
]

# Long condition — split across lines inside ()
port    = 23
service = "Telnet"

is_high_risk = (
    port in dangerous_ports   # port must be on the dangerous list
    and service == "Telnet"   # AND service must be Telnet specifically
)

print(f"Port {port} high risk: {is_high_risk}")
print(f"Dangerous ports list: {dangerous_ports}")
````
