## 1. What is a Variable?

**Analogy:** Think of a variable as a labelled box on a shelf. The **label** is the variable name. The **contents** are the value. When you need the value later, you just refer to the label — you do not need to remember what is inside.

```
+-----------------------+     +-----------------------+     +-----------------------+
|  Label: target_ip     |     |  Label: cvss_score    |     |  Label: is_patched    |
|-----------------------|     |-----------------------|     |-----------------------|
|  Contents: 10.0.0.55  |     |  Contents: 9.8        |     |  Contents: False      |
+-----------------------+     +-----------------------+     +-----------------------+
```

To **create** a variable in Python, you write the label, then `=`, then the value:

```python
label = value
```

The `=` sign in Python does **not** mean "equal to" (that is `==`). It means **"store this value"**. Read it as: *"target_ip stores the value 10.0.0.55"*.

Once stored, you can use the variable name anywhere you would use the value directly.


````python
# Creating variables
target_ip    = "10.0.0.55"   # stores a text value
cvss_score   = 9.8           # stores a decimal number
is_patched   = False         # stores a True/False flag

# Use the variable name to access the value
target_ip    # Jupyter shows the last expression's value automatically
````

### Variables can be updated

A variable can hold a new value at any time. The old value is simply replaced.


````python
# Variables can be reassigned
alert_level = "LOW"
alert_level = "CRITICAL"   # old value "LOW" is gone

alert_level
````

## 2. Naming Rules

Python has strict rules for variable names. Break them and you get a `SyntaxError`. Follow them and your code stays readable.

| Rule | Allowed | Not allowed |
|------|---------|-------------|
| Start with a letter or `_` | `target_ip`, `_private` | `1port`, `@host` |
| Only letters, digits, `_` after that | `failed_attempts`, `host2` | `host-name`, `my ip` |
| No spaces | `scan_result` | `scan result` |
| Not a Python keyword | `port`, `hostname` | `for`, `if`, `class` |

### Convention: snake_case

Python's style guide (PEP 8) says variable names use **snake_case**: all lowercase, words separated by underscores.

```python
# Bad naming
x         = 9.8          # meaningless — what is x?
CVSS      = 9.8          # looks like a constant
cvssScore = 9.8          # camelCase — that is Java style, not Python

# Good naming
cvss_score = 9.8         # snake_case, descriptive
```

Constants (values that never change) use `UPPER_SNAKE_CASE`:

```python
MAX_LOGIN_ATTEMPTS = 3
DEFAULT_SSH_PORT   = 22
```


````python
# Good names: descriptive, snake_case, security context
failed_attempts   = 7
blocked_ip        = "203.0.113.42"
firewall_enabled  = True
MAX_RETRIES       = 3    # constant: value never changes

# These would cause errors (commented out to prevent crash)
# 1st_attempt = 1       # SyntaxError: starts with a digit
# blocked-ip  = "..."   # SyntaxError: hyphen not allowed
# for         = 5       # SyntaxError: reserved keyword

blocked_ip
````

## 3. Data Types

Every value in Python has a **type**. The type tells Python what the value is and what you can do with it. You do not have to declare the type — Python figures it out from the value you assign.

| Type | Name | Example values |
|------|------|----------------|
| `str` | String | `"admin"`, `"192.168.1.1"`, `"CVE-2024-1234"` |
| `int` | Integer | `22`, `443`, `3`, `-1` |
| `float` | Float | `9.8`, `7.5`, `0.0` |
| `bool` | Boolean | `True`, `False` |

### Strings — text values

Wrap text in single quotes `'...'` or double quotes `"..."`. Either works — just be consistent.

Even if a value looks like a number, putting it in quotes makes it a string:

```python
port_number = 443          # int  — you can do maths with this
port_string = "443"        # str  — you cannot add 1 to this directly
```


````python
# Strings: anything in quotes
analyst_name    = "Alice Dupont"
target_hostname = "db-server-01"
cve_id          = "CVE-2024-1234"
log_entry       = "FAILED login from 192.168.1.100 at 03:47:22"

# Single or double quotes — both valid
protocol        = 'HTTPS'

cve_id
````

### Integers — whole numbers

No quotes, no decimal point. Use integers for counting, port numbers, IDs, and anything you add, subtract, or multiply.


````python
# Integers: whole numbers, no quotes
ssh_port         = 22
https_port       = 443
failed_attempts  = 5
alert_count      = 0

# Integers support arithmetic
total_attempts = failed_attempts + alert_count

total_attempts
````

### Floats — decimal numbers

Use floats for CVSS scores, percentages, response times, and any value that has a fractional part.


````python
# Floats: decimal numbers
cvss_score        = 9.8    # CRITICAL severity
scan_duration_sec = 12.4
cpu_usage_pct     = 78.5
risk_level        = 0.0    # 0.0 is still a float, not an int

cvss_score
````

### Booleans — True or False

A boolean stores exactly one of two values: `True` or `False`. The capital letter is mandatory — `true` and `false` are not valid Python.

Use booleans for flags and status indicators: is the firewall active? Is the host reachable? Is the account locked?


````python
# Booleans: True or False only
firewall_enabled = True
vpn_connected    = False
account_locked   = True
patch_applied    = False

# Common mistake: lowercase does not work
# firewall_enabled = true   # NameError: name 'true' is not defined

account_locked
````

### type() — Checking what type a variable holds

`type()` is a built-in function that returns the type of any value. Use it whenever you are unsure what you are working with.


````python
# type() tells you what Python thinks a variable is
print(type("192.168.1.1"))   # -> <class 'str'>
print(type(443))              # -> <class 'int'>
print(type(9.8))              # -> <class 'float'>
print(type(True))             # -> <class 'bool'>

# Works on variables too
cvss_score = 9.8
print(type(cvss_score))       # -> <class 'float'>
````

````python
# Classic trap: looks like a number, is actually a string
port_a = 443          # int  — no quotes
port_b = "443"        # str  — has quotes

print(type(port_a))   # <class 'int'>
print(type(port_b))   # <class 'str'>

# They look the same when printed, but are not the same type
print(port_a, "vs", port_b)
print(port_a == port_b)    # False! int 443 != str "443"
````

## 4. print() - Displaying Your Variables

**Analogy:** A variable is information stored silently in memory. `print()` is your way of speaking it out loud — it sends the value to the screen so you can see it. Every analyst, developer, and programmer uses `print()` dozens of times a day to check what is happening inside a script.

```python
print(what_you_want_to_see)
```

You can pass:
- A literal value: `print("Scan complete")`
- A variable: `print(target_ip)`
- Multiple things separated by commas: `print("Port:", port_number)`


````python
# print() basics
print("Scan complete")           # print a text string
print(9.8)                       # print a number
print(True)                      # print a boolean

# print a variable
target_ip = "10.0.0.55"
print(target_ip)                 # prints the value, not the name
````

### Printing multiple values

Separate multiple items with commas. Python automatically adds a space between them.


````python
# Multiple values in one print() call
target_ip   = "10.0.0.55"
open_port   = 22
service     = "SSH"

print("Target:", target_ip)                  # label + variable
print("Port:", open_port, "Service:", service) # mix text and variables
print(target_ip, open_port, service)          # just variables
````

### sep and end — controlling the output

By default, `print()` puts a space between items and a newline `\n` at the end. You can change both:

- `sep="..."` — what to put **between** items (default: `" "`)
- `end="..."` — what to put **at the end** (default: `"\n"` — a newline)


````python
# sep: change separator between items
print("10", "0", "0", "55", sep=".")      # -> 10.0.0.55 (looks like an IP)
print("DENY", "src=1.2.3.4", "port=22", sep=" | ")  # -> DENY | src=1.2.3.4 | port=22

# end: stay on the same line
print("[*] Scanning", end=" ")   # no newline
print("10.0.0.55")                # continues on the same line

# end with nothing: completely suppress the newline
print("RESULT:", end="")
print(" PASS")
````

## 5. input() - Reading from the User

**Analogy:** `input()` is like a form on a screen. You display a prompt ("Enter the target IP:"), the user types something, and Python stores whatever they typed in a variable.

```python
variable = input("Your prompt text here: ")
```

The program **pauses** at `input()` and waits. Once the user presses Enter, the typed text is stored in the variable and execution continues.


````python
# Basic input() — ask the user for a value and store it
# When you run this cell, type something and press Enter

analyst_name = input("Enter your analyst name: ")
print("Logged in as:", analyst_name)
````

### Warning - input() always returns a string

This is the most common `input()` mistake. **Whatever the user types, Python stores it as a string** — even if they typed a number.


````python
# Demonstrate: input() always returns str
# Run this and type  443  then Enter
raw_port = input("Enter a port number: ")

print("You typed:", raw_port)
print("Type:", type(raw_port))   # -> <class 'str'> even though it looks like a number
````

### Converting input to the right type

Because `input()` always returns a string, you must convert it if you need a number or boolean. Use `int()` or `float()` to wrap the `input()` call:

```python
port = int(input("Enter port: "))     # converts string "443" -> integer 443
score = float(input("Enter score: ")) # converts string "9.8" -> float 9.8
```


````python
# Pattern: convert input immediately at the point of reading
# Run this and type a number like  22  then Enter
port_number = int(input("Enter the target port: "))

print("Port:", port_number)
print("Type:", type(port_number))   # -> <class 'int'>
print("Port + 1 =", port_number + 1)  # arithmetic works now
````

````python
# A small security questionnaire using input()
# Run this cell and answer both questions

target_host = input("Target hostname or IP: ")
scan_depth  = int(input("Scan depth (1-5): "))

print()
print("--- Scan Configuration ---")
print("Host:", target_host)
print("Depth:", scan_depth)
print("Type check - host:", type(target_host), "| depth:", type(scan_depth))
````

## 6. String Formatting

Once you have values in variables, you frequently need to build **messages** that combine text and variable values. Python provides several ways to do this.

### Method 1: f-strings (recommended — Python 3.6+)

Put an `f` before the opening quote, then place variable names inside `{}`. Python replaces each `{}` with the variable's value automatically.

```python
f"Text {variable_name} more text"
```

Read the `f` as: *"fill in the blanks from the variables"*.


````python
# f-strings: clean, readable, handles any type automatically
target_ip    = "10.0.0.55"
open_port    = 443
cvss_score   = 9.8
patch_status = False

# Build a report line
report = f"Host {target_ip} has port {open_port} open — CVSS {cvss_score} — patched: {patch_status}"
print(report)

# Expressions work inside {}
total_ports = 65535
scanned     = 1000
print(f"Coverage: {scanned / total_ports * 100:.1f}%")  # :.1f rounds to 1 decimal
````

### Method 2: .format() (older code — you will encounter this)

Place `{}` placeholders in the string, then call `.format()` with the values. You will see this style in older code and tutorials.


````python
# .format(): placeholders filled in order by position
target_ip = "10.0.0.55"
open_port = 443

message = "Scanning {} on port {}".format(target_ip, open_port)
print(message)

# Named placeholders: more readable for longer strings
alert = "[{severity}] CVE {cve} detected on {host}".format(
    severity="CRITICAL",
    cve="2024-1234",
    host=target_ip
)
print(alert)
````

### Which to use?

| Situation | Use |
|---|---|
| Any new code you write | f-string (`f"...{var}..."`) |
| Reading or maintaining old code | `.format()` — understand it, no need to rewrite |
| Never use in new code | `%s` style (Python 2 legacy) |

## 7. Type Conversion

You will frequently need to convert a value from one type to another. The most common case is converting string input from `input()` into an integer or float.

| Function | What it does | Example |
|----------|-------------|--------|
| `int(x)` | Converts x to integer | `int("443")` -> `443` |
| `float(x)` | Converts x to float | `float("9.8")` -> `9.8` |
| `str(x)` | Converts x to string | `str(443)` -> `"443"` |
| `bool(x)` | Converts x to boolean | `bool(0)` -> `False` |


````python
# int() and float(): convert string numbers to numeric types
port_str  = "443"         # came from a config file as text
score_str = "9.8"         # same

port_int   = int(port_str)
score_float = float(score_str)

print(port_int + 1)       # 444 — arithmetic now works
print(score_float * 2)    # 19.6

# str(): convert numbers back to text for building messages
port_label = "Port: " + str(port_int)   # concatenation requires both sides to be str
print(port_label)
````

````python
# bool() conversion rules:
# False: 0, 0.0, "", None, and empty collections
# True: everything else

print(bool(0))          # False — zero is falsy
print(bool(1))          # True  — any non-zero number is truthy
print(bool(""))         # False — empty string is falsy
print(bool("admin"))    # True  — non-empty string is truthy
print(bool(5))          # True  — failed_attempts > 0 means something happened
````

### Warning - Conversion failures

`int()` and `float()` will **crash** if the string cannot be converted. Handle this carefully:

```python
int("22")        # OK -> 22
int("22.5")      # ValueError: "22.5" has a decimal — use float() instead
int("PORT-22")   # ValueError: cannot convert text to int
```

You will learn how to handle errors safely in the Exception Handling module.

## 8. Summary

| Concept | Key rule |
|---------|----------|
| **Variable** | A labelled container. Created with `name = value`. Reassignable at any time. |
| **Naming** | `snake_case`. Starts with letter or `_`. No spaces, no hyphens, no keywords. Descriptive. |
| **str** | Text in quotes. IP addresses, names, log entries, IDs. |
| **int** | Whole numbers. Ports, counts, IDs. No quotes. |
| **float** | Decimal numbers. CVSS scores, percentages. No quotes. |
| **bool** | `True` or `False` only. Capital first letter required. |
| **type()** | Returns the type of any value. Useful for debugging. |
| **print()** | Displays values to the screen. Comma-separates multiple items. `sep=` and `end=` control formatting. |
| **input()** | Reads text from the user. **Always returns a string.** Wrap in `int()` or `float()` for numbers. |
| **f-strings** | `f"...{variable}..."` — embed variables in text. The preferred method. |
| **Type conversion** | `int()`, `float()`, `str()`, `bool()`. Crashes if conversion is impossible. |
