## Testing

Run this cell to check all 15 exercises automatically. Read the error message for any test that fails — it will tell you exactly what went wrong.

> The `class TestVariables(unittest.TestCase):` syntax uses OOP concepts covered in a later module. Do not worry about understanding it now — just run the cell.


````python
import unittest


class TestVariables(unittest.TestCase):

    # Part A
    def test_01_server_hostname(self):
        self.assertEqual(server_hostname, "firewall-01",
                         "server_hostname should be the string 'firewall-01'")
        self.assertIsInstance(server_hostname, str,
                              "server_hostname must be a str, not an int or other type")

    def test_02_https_port(self):
        self.assertEqual(https_port, 443,
                         "https_port should be the integer 443 (no quotes)")
        self.assertIsInstance(https_port, int,
                              "https_port must be an int, not a string")

    def test_03_is_encrypted(self):
        self.assertEqual(is_encrypted, True,
                         "is_encrypted should be True (capital T, no quotes)")
        self.assertIsInstance(is_encrypted, bool,
                              "is_encrypted must be a bool, not an int")

    # Part B
    def test_04_cvss_score_and_type(self):
        self.assertEqual(cvss_score, 8.1,
                         "cvss_score should be the float 8.1")
        self.assertIsInstance(cvss_score, float,
                              "cvss_score must be a float, not an int")
        self.assertEqual(cvss_type, float,
                         "cvss_type should be the type 'float' — use type(cvss_score)")

    def test_05_source_ip_and_type(self):
        self.assertEqual(source_ip, "203.0.113.42",
                         "source_ip should be '203.0.113.42'")
        self.assertEqual(ip_type, str,
                         "ip_type should be the type 'str' — IP addresses are strings")

    def test_06_total_alerts(self):
        self.assertEqual(total_alerts, 5,
                         "total_alerts should be 5 — convert alerts_today from str to int first")
        self.assertIsInstance(total_alerts, int,
                              "total_alerts must be an int")

    # Part C
    def test_08_scan_header(self):
        self.assertEqual(scan_header, "[SCAN] Target: 10.0.0.55 | Ports: 1000",
                         "scan_header must exactly match '[SCAN] Target: 10.0.0.55 | Ports: 1000'")

    def test_09_report_line(self):
        self.assertEqual(report_line, "CVE-2024-5678 | CVSS: 9.8 | CRITICAL",
                         "report_line must exactly match 'CVE-2024-5678 | CVSS: 9.8 | CRITICAL' — use :.1f")

    # Part D
    def test_10_type_conversions(self):
        self.assertEqual(port_int, 22,
                         "port_int should be the integer 22, not the string '22'")
        self.assertIsInstance(port_int, int,
                              "port_int must be an int")
        self.assertEqual(score_float, 7.5,
                         "score_float should be the float 7.5")
        self.assertIsInstance(score_float, float,
                              "score_float must be a float")
        self.assertEqual(count_int, 12,
                         "count_int should be the integer 12")

    def test_11_strip_and_convert(self):
        self.assertEqual(clean_port, 8080,
                         "clean_port should be integer 8080 — use dirty_port.strip() before int()")
        self.assertIsInstance(clean_port, int,
                              "clean_port must be an int")
        self.assertEqual(clean_score, 9.8,
                         "clean_score should be float 9.8 — use dirty_score.strip() before float()")
        self.assertIsInstance(clean_score, float,
                              "clean_score must be a float")

    def test_12_simulated_input(self):
        self.assertEqual(username_input, "alice",
                         "username_input should be the string 'alice'")
        self.assertEqual(attempts_input, "4",
                         "attempts_input should be the string '4' (as input() would return)")
        self.assertEqual(failed_logins, 4,
                         "failed_logins should be the integer 4 — convert attempts_input with int()")
        self.assertIsInstance(failed_logins, int,
                              "failed_logins must be an int, not a string")

    # Part E
    def test_13_incident_record(self):
        self.assertEqual(incident_ref, "INC-2024-0099",
                         "incident_ref should be the string 'INC-2024-0099'")
        self.assertEqual(attacker_ip, "185.220.101.5",
                         "attacker_ip should be the string '185.220.101.5'")
        self.assertEqual(target_port, 3306,
                         "target_port should be the integer 3306 — MySQL port")
        self.assertIsInstance(target_port, int,
                              "target_port must be an int, not a string")
        self.assertEqual(risk_score, 9.1,
                         "risk_score should be the float 9.1")
        self.assertIsInstance(risk_score, float,
                              "risk_score must be a float")
        self.assertEqual(is_isolated, False,
                         "is_isolated should be False")
        self.assertIsInstance(is_isolated, bool,
                              "is_isolated must be a bool")

    def test_14_incident_summary(self):
        self.assertEqual(
            incident_summary,
            "[INC-2024-0099] 185.220.101.5:3306 | risk: 9.1 | isolated: False",
            "incident_summary must exactly match the required format string"
        )

    def test_15_total_failed_ssh(self):
        self.assertEqual(total_failed_ssh, 60,
                         "total_failed_ssh should be 60 (14 + 37 + 9)")
        self.assertIsInstance(total_failed_ssh, int,
                              "total_failed_ssh must be an int")


unittest.main(argv=[""], verbosity=2, exit=False)
````
