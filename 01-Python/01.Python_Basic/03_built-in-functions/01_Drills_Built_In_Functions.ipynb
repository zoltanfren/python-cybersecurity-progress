## Testing

Run the cell below to check all 15 exercises at once.
Read each error message carefully — it tells you exactly what was expected and what you produced.

Do not modify this cell.

````python
import unittest


class TestBuiltInFunctions(unittest.TestCase):

    # Part 1 — type() and type conversion

    def test_01_count_alpha(self):
        self.assertIsInstance(count_alpha, int,
            "count_alpha must be an int — len() returns an integer")
        self.assertEqual(count_alpha, 12,
            "count_alpha should be 12: len('Hello world!') = 12 characters including the space and !")

    def test_02_count_float(self):
        self.assertIsInstance(count_float, float,
            "count_float must be a float — use float(count_alpha)")
        self.assertEqual(count_float, 12.0,
            "count_float should be 12.0: float(12) = 12.0")

    def test_03_port_conversion(self):
        self.assertIsInstance(port_int, int,
            "port_int must be an int — use int('443')")
        self.assertEqual(port_int, 443,
            "port_int should be 443: int('443') = 443")
        self.assertIsInstance(is_privileged, bool,
            "is_privileged must be a bool, not an int (use < not -1)")
        self.assertEqual(is_privileged, True,
            "is_privileged should be True: 443 < 1024")

    # Part 2 — round(), abs(), bool()

    def test_04_round_pi(self):
        self.assertIsInstance(round_pi, float,
            "round_pi must be a float")
        self.assertEqual(round_pi, 3.14,
            "round_pi should be 3.14: round(math.pi, 2) = 3.14")

    def test_05_score_gap(self):
        self.assertIsInstance(score_gap, float,
            "score_gap must be a float (result of abs on two floats)")
        self.assertAlmostEqual(score_gap, 3.5, places=5,
            msg="score_gap should be 3.5: abs(9.8 - 6.3) = 3.5")

    def test_06_firewall_enabled(self):
        self.assertIsInstance(firewall_enabled, bool,
            "firewall_enabled must be a bool — use bool(firewall_flag)")
        self.assertEqual(firewall_enabled, True,
            "firewall_enabled should be True: bool(1) = True")

    # Part 3 — min(), max(), sum()

    def test_07_sum_of_list(self):
        self.assertIsInstance(sum_of_list, int,
            "sum_of_list must be an int (sum of integers is an integer)")
        self.assertEqual(sum_of_list, 31,
            "sum_of_list should be 31: 2+8+1+4+6+3+7 = 31")

    def test_08_min_max(self):
        self.assertEqual(min_value, 1,
            "min_value should be 1: the smallest value in [2,8,1,4,6,3,7]")
        self.assertEqual(max_value, 8,
            "max_value should be 8: the largest value in [2,8,1,4,6,3,7]")

    def test_09_cvss_min_max(self):
        self.assertAlmostEqual(lowest_cvss, 4.0, places=5,
            msg="lowest_cvss should be 4.0: min(9.8, 4.0, 7.5) = 4.0")
        self.assertAlmostEqual(highest_cvss, 9.8, places=5,
            msg="highest_cvss should be 9.8: max(9.8, 4.0, 7.5) = 9.8")

    # Part 4 — sorted(), reversed()

    def test_10_sorted_num(self):
        self.assertIsInstance(sorted_num, list,
            "sorted_num must be a list — sorted() returns a list")
        self.assertEqual(sorted_num, [1, 2, 3, 4, 6, 7, 8],
            "sorted_num should be [1,2,3,4,6,7,8]: sorted([2,8,1,4,6,3,7]) in ascending order")

    def test_11_reversed_text(self):
        self.assertIsInstance(reversed_text, list,
            "reversed_text must be a list — wrap reversed() with list()")
        self.assertEqual(reversed_text,
            ['!', 'd', 'l', 'r', 'o', 'w', ' ', 'o', 'l', 'l', 'e', 'H'],
            "reversed_text should be list(reversed('Hello world!'))")

    def test_12_triage_order(self):
        self.assertIsInstance(triage_order, list,
            "triage_order must be a list")
        self.assertEqual(triage_order, [9.8, 7.5, 5.5, 4.0, 2.1],
            "triage_order should be [9.8, 7.5, 5.5, 4.0, 2.1]: sorted descending for triage")

    # Part 5 — Putting It Together

    def test_13_eval(self):
        self.assertIsInstance(string_interpret, int,
            "string_interpret must be an int — eval('1 + 2') returns 3, an integer")
        self.assertEqual(string_interpret, 3,
            "string_interpret should be 3: eval('1 + 2') = 3")

    def test_14_port_range(self):
        self.assertIsInstance(port_range, list,
            "port_range must be a list — use list(range(...))")
        self.assertEqual(port_range, [8000, 8001, 8002, 8003, 8004, 8005],
            "port_range should be [8000..8005]: list(range(8000, 8006))")

    def test_15_cvss_pipeline(self):
        self.assertIsInstance(cvss_score, float,
            "cvss_score must be a float — use float(raw_cvss)")
        self.assertEqual(cvss_score, 7.0,
            "cvss_score should be 7.0: float('7') = 7.0")
        self.assertAlmostEqual(cvss_deviation, 2.0, places=5,
            msg="cvss_deviation should be 2.0: abs(7.0 - 5.0) = 2.0")
        self.assertAlmostEqual(cvss_deviation_rounded, 2.0, places=1,
            msg="cvss_deviation_rounded should be 2.0: round(2.0, 1) = 2.0")


unittest.main(argv=[""], verbosity=2, exit=False)
````
