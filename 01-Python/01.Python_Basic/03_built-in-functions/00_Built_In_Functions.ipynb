## 1. What Are Built-in Functions?

### Your hands vs your toolbox

Think of Python's built-in functions as your **bare hands** — tools you always have with you, no preparation needed. Other functions require you to first fetch a tool from a toolbox (`import`). Built-ins are just always there.

### How to call a function

```python
result = function_name(argument)
```

- **`function_name`** — the name of the function
- **`argument`** — the value you pass in (what you want the function to work on)
- **`result`** — what the function gives back (the return value)

### Built-ins covered in this notebook

| Function | What it does |
|---|---|
| `type()` | Returns the type of a value |
| `int()` | Converts to integer |
| `float()` | Converts to float |
| `str()` | Converts to string |
| `bool()` | Converts to boolean |
| `len()` | Returns the length of a string or sequence |
| `abs()` | Returns the absolute value |
| `round()` | Rounds to a given number of decimal places |
| `min()` | Returns the smallest value |
| `max()` | Returns the largest value |
| `sum()` | Returns the total of all values |
| `sorted()` | Returns a sorted copy |
| `reversed()` | Returns items in reverse order |
| `range()` | Generates a sequence of integers |
| `input()` | Reads text typed by the user |
| `eval()` | Evaluates a string as Python code |


<br>

## 2. `type()` — Inspect Any Value

### What does `type()` do?

`type()` tells you **what kind of value** you are dealing with. It returns the type object.

```python
type(value)  →  <class 'int'> / <class 'str'> / <class 'float'> / <class 'bool'>
```

### Why it matters

User input always arrives as a **string**, even if it looks like a number. Before using a value in a calculation (like comparing a CVSS score), you must know — and often change — its type.

````python
# type(): check the type of different values

cvss_score     = 9.8
port_number    = 443
protocol_name  = "HTTPS"
firewall_on    = True

print(type(cvss_score))    # <class 'float'>
print(type(port_number))   # <class 'int'>
print(type(protocol_name)) # <class 'str'>
print(type(firewall_on))   # <class 'bool'>
````

````python
# type() in a conditional: only process the value if it is the right type
# Input from a user or a config file might arrive with the wrong type

raw_score = "9.8"  # arrived as a string from user input

if type(raw_score) == str:
    print(f"Warning: score is a string, not a number. Convert before use.")
else:
    print(f"CVSS score: {raw_score}")
````

## 3. Type Conversion: `int()`, `float()`, `str()`, `bool()`

### What is type conversion?

Type conversion (also called **casting**) transforms a value from one type into another.

| Function | Converts to | Example | Result |
|---|---|---|---|
| `int()` | Integer | `int("443")` | `443` |
| `int()` | Integer (drops decimal) | `int(9.8)` | `9` |
| `float()` | Float | `float("9.8")` | `9.8` |
| `float()` | Float | `float(443)` | `443.0` |
| `str()` | String | `str(443)` | `"443"` |
| `bool()` | Boolean | `bool(0)` | `False` |
| `bool()` | Boolean | `bool(1)` | `True` |

### ⚠️ The most important rule: `input()` always returns a string

If you ask the user to type a port number, Python gives you `"443"` (a string), not `443` (an integer). You **must** convert it before doing arithmetic or comparisons with numbers.

````python
# The input() problem: a port number arrives as a string
# Without conversion, comparisons with integers fail or behave unexpectedly

port_as_string = "443"  # simulates what input() would return
port_as_int    = int(port_as_string)  # convert to integer for proper use

print(type(port_as_string))  # <class 'str'>
print(type(port_as_int))     # <class 'int'>
print(port_as_int > 1024)    # False — this comparison works correctly now
````

````python
# Casting chain: simulate receiving a CVSS score as text and converting it

raw_input    = "9"            # string from user
score_int    = int(raw_input) # string -> integer
score_float  = float(raw_input) # string -> float
score_back   = str(score_int)   # integer -> string (for building a message)

print(f"As string : {raw_input!r}  type: {type(raw_input)}")
print(f"As int    : {score_int}     type: {type(score_int)}")
print(f"As float  : {score_float}   type: {type(score_float)}")
print(f"Back to str: {score_back!r} type: {type(score_back)}")
````

````python
# bool(): any non-zero integer is True; zero is False
# Useful when interpreting numeric flags from config files or network packets

firewall_flag = 1  # 1 = enabled, 0 = disabled (common in config files)
firewall_on   = bool(firewall_flag)

print(f"Firewall enabled: {firewall_on}")  # True

disabled_flag = 0
print(f"Firewall enabled: {bool(disabled_flag)}")  # False
````

## 4. `len()` — Measure Length

### What does `len()` do?

`len()` returns the **number of items** in a string (characters) or a sequence (elements).

```python
len("HTTPS")    # 5 — number of characters
len([80, 443])  # 2 — number of elements
```

### Security use cases

- Enforcing a **minimum password length** policy
- Checking if a CVE ID has the expected format length
- Counting how many open ports or findings a scan returned

````python
# len() on strings: password length enforcement
# A common policy: passwords must be at least 12 characters

password       = "Adm1n!Secure#"
min_length     = 12
password_length = len(password)

print(f"Password length: {password_length}")

if password_length >= min_length:
    print("Password meets length policy")
else:
    print(f"Password too short: need {min_length - password_length} more character(s)")
````

````python
# len() on a CVE ID to validate expected format
# A standard CVE ID 'CVE-YYYY-NNNN' has exactly 12 characters for 4-digit years/IDs

cve_id = "CVE-2024-1234"
print(f"CVE ID length: {len(cve_id)}")
````

## 5. Math Helpers: `abs()`, `round()`, `min()`, `max()`, `sum()`

### `abs()` — Absolute value

`abs()` removes the sign from a number. Negative becomes positive; positive stays positive.

```python
abs(-7.5)  # 7.5
abs(7.5)   # 7.5
```

### `round()` — Round to N decimal places

```python
round(9.8765, 2)   # 9.88
round(9.8765, 0)   # 10.0
round(9.8765)      # 10  (no second argument: rounds to nearest integer)
```

### `min()`, `max()`, `sum()` — Aggregate functions

These three functions can take **individual arguments** or **a sequence** (like a list):

```python
min(9, 7, 4)           # 4  — individual arguments
max([9, 7, 4])         # 9  — a list
sum([9, 7, 4])         # 20 — a list
```

````python
# abs(): calculate the score deviation from the average
# Useful when comparing a specific finding to a baseline

average_cvss = 6.5
finding_cvss = 4.0

deviation = abs(finding_cvss - average_cvss)  # always positive regardless of direction
print(f"Score deviation from average: {deviation}")
````

````python
# round(): display a CVSS score to 2 decimal places

import math

raw_pi  = math.pi         # 3.141592653589793
round_pi = round(raw_pi, 2)

print(f"Full value : {raw_pi}")
print(f"Rounded to 2 dp: {round_pi}")
````

````python
# min(), max(), sum() with individual values: summarise a set of CVSS scores

lowest_score  = min(9.8, 7.5, 4.0, 5.5)
highest_score = max(9.8, 7.5, 4.0, 5.5)
total_score   = sum([9.8, 7.5, 4.0, 5.5])  # sum() requires a sequence

print(f"Lowest CVSS  : {lowest_score}")
print(f"Highest CVSS : {highest_score}")
print(f"Total        : {total_score}")
````

## 6. `sorted()` and `reversed()`

### `sorted()` — Return a sorted copy

`sorted()` takes a sequence and **returns a new sorted list**. The original is not modified.

```python
sorted([3, 1, 2])              # [1, 2, 3]  ascending (default)
sorted([3, 1, 2], reverse=True) # [3, 2, 1]  descending
```

### `reversed()` — Iterate in reverse order

`reversed()` returns a **reverse iterator** — not a list directly. To get a list, wrap it with `list()`:

```python
list(reversed([1, 2, 3]))      # [3, 2, 1]
list(reversed("ABC"))          # ['C', 'B', 'A']
```

### `sorted()` vs `reversed()`

| Function | What it does | Returns |
|---|---|---|
| `sorted()` | Sorts by value | New list, always ordered |
| `reversed()` | Flips the order | Iterator (wrap with `list()` to use) |

````python
# sorted(): rank CVSS scores from highest to lowest for triage prioritisation

cvss_scores  = [4.0, 9.8, 7.5, 5.5, 2.1]
sorted_asc   = sorted(cvss_scores)              # lowest first
sorted_desc  = sorted(cvss_scores, reverse=True) # highest first (triage order)

print(f"Original  : {cvss_scores}")
print(f"Ascending : {sorted_asc}")
print(f"Descending: {sorted_desc}")
````

````python
# reversed() on a string: reverse a password for simple analysis
# Returns an iterator — list() materialises it into a usable list

original_text  = "Hello world!"
reversed_text  = list(reversed(original_text))

print(f"Original : {original_text}")
print(f"Reversed : {reversed_text}")
````

## 7. `range()` — Generate Number Sequences

### What is `range()`?

`range()` generates a sequence of integers. It does not create a list in memory — it produces numbers one at a time, which is memory-efficient. You will use it constantly in the **Loops module** that follows this one.

### Three forms

| Form | Produces | Example |
|---|---|---|
| `range(stop)` | 0 up to (not including) stop | `range(5)` → 0, 1, 2, 3, 4 |
| `range(start, stop)` | start up to (not including) stop | `range(1, 5)` → 1, 2, 3, 4 |
| `range(start, stop, step)` | start to stop, jumping by step | `range(0, 10, 2)` → 0, 2, 4, 6, 8 |

### ⚠️ stop is always exclusive

`range(1, 6)` gives you 1, 2, 3, 4, 5 — it stops **before** 6.

### Seeing the values

To inspect what a `range()` produces, wrap it with `list()`:

````python
# Inspect range() output by converting to a list
# In real use you will not call list() — you iterate directly with a for loop

print(list(range(5)))            # [0, 1, 2, 3, 4]
print(list(range(1, 6)))         # [1, 2, 3, 4, 5]
print(list(range(80, 91, 10)))   # [80, 90]  — every 10th port
````

````python
# range() for a port scan: ports 1 through 10
# list() is used here only to show the output — in the Loops module you will iterate directly

scan_range = list(range(1, 11))
print(f"Ports to scan: {scan_range}")
````

## 8. `input()` — Read User Input

### What does `input()` do?

`input()` pauses the program, displays a prompt, and waits for the user to type something and press Enter.

```python
username = input("Enter your username: ")
```

### The critical rule: `input()` always returns a string

Even if the user types `443`, Python gives you the **string** `"443"` — not the integer `443`. Always convert immediately after receiving the input.

```python
port_number = int(input("Enter port number: "))  # convert on the same line
```

````python
# input(): ask for a username and display it
# Note: this cell is interactive — it will pause and wait for you to type

username = input("Enter your username: ")
print(f"Hello, {username}")
print(f"Type of input: {type(username)}")  # always <class 'str'>
````

````python
# input() with immediate conversion: port number as integer
# int() wraps input() so the result is already an integer

port_number = int(input("Enter a port number: "))
print(f"Port entered: {port_number}")
print(f"Type: {type(port_number)}")  # <class 'int'>

if port_number < 1024:
    print("Privileged port (0–1023)")
else:
    print("Unprivileged port (1024+)")
````

## 9. `eval()` — Power and Danger

### What does `eval()` do?

`eval()` takes a **string** and executes it as Python code, returning the result.

```python
eval("1 + 2")      # 3
eval("9 > 7")      # True
```

### ⚠️ Security warning: `eval()` is dangerous

**Never use `eval()` on input you do not fully control.** If a user can control the string passed to `eval()`, they can execute arbitrary Python code on your system:

```python
user_input = "__import__('os').system('rm -rf /')"  # deletes the filesystem
eval(user_input)  # DO NOT do this
```

This is a real vulnerability class called **code injection**.

### When `eval()` is acceptable

Only on **fully controlled, hardcoded strings** — never on user input. In most real-world scenarios, there is a safer alternative.

````python
# eval(): interpret a hardcoded arithmetic string expression
# This is acceptable ONLY because we control the string

calc             = "1 + 2"
string_interpret = eval(calc)

print(f"Expression : {calc!r}")
print(f"Result     : {string_interpret}")
print(f"Type       : {type(string_interpret)}")
````

## Summary

| Function | Purpose | Example |
|---|---|---|
| `type(x)` | Returns the type of `x` | `type(443)` → `<class 'int'>` |
| `int(x)` | Converts `x` to integer | `int("443")` → `443` |
| `float(x)` | Converts `x` to float | `float("9.8")` → `9.8` |
| `str(x)` | Converts `x` to string | `str(443)` → `"443"` |
| `bool(x)` | Converts `x` to boolean | `bool(0)` → `False` |
| `len(x)` | Length of `x` | `len("HTTPS")` → `5` |
| `abs(x)` | Absolute value | `abs(-3.5)` → `3.5` |
| `round(x, n)` | Round to `n` decimals | `round(3.14159, 2)` → `3.14` |
| `min(...)` | Smallest value | `min(9, 7, 4)` → `4` |
| `max(...)` | Largest value | `max(9, 7, 4)` → `9` |
| `sum(seq)` | Total of all values | `sum([9, 7, 4])` → `20` |
| `sorted(seq)` | Returns sorted copy | `sorted([3,1,2])` → `[1, 2, 3]` |
| `reversed(seq)` | Returns reverse iterator | `list(reversed([1,2,3]))` → `[3, 2, 1]` |
| `range(start, stop, step)` | Generates integers | `list(range(1, 4))` → `[1, 2, 3]` |
| `input(prompt)` | Reads user input | `input("Name: ")` → `"Alice"` |
| `eval(string)` | Executes string as code | `eval("1+2")` → `3` |

### Key rules to remember

- `input()` **always** returns a string — convert before arithmetic
- `int()` truncates floats: `int(9.9)` → `9`, not `10`
- `sorted()` returns a **new list** — the original is unchanged
- `reversed()` returns an **iterator** — wrap with `list()` to materialise
- `range()` stop is **exclusive**: `range(1, 5)` → 1, 2, 3, 4
- **Never** use `eval()` on user input
