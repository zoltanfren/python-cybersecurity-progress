## Part 1 — `for` Loops with `range()`

### Exercise 1

Create a variable `port_total` starting at `0`.

Expected result: `port_total == 15`

````python
# Exercise 1
````

<details>
<summary><span style="color:#e74c3c; font-size:1.1em;"><strong>💡 SOLUTION</strong></span></summary>

## Part 5 — Putting It Together

### Exercise 13

Create a variable `total_digits` starting at `0`.
Iterate over the string `"CVE-2024-1234"` character by character.
For each character that is a digit, add its **integer value** (`int(char)`) to `total_digits`.

Expected result: `total_digits == 16`  (2+0+2+4+1+2+3+4 = 18... wait: 2+0+2+4+1+2+3+4 = 18)

Expected result: `total_digits == 18`

````python
# Exercise 13
````

<details>
<summary><span style="color:#e74c3c; font-size:1.1em;"><strong>💡 SOLUTION</strong></span></summary>

```python
# SOLUTION
# Digits in 'CVE-2024-1234': 2, 0, 2, 4, 1, 2, 3, 4
# Sum: 2+0+2+4+1+2+3+4 = 18
total_digits = 0
for character in "CVE-2024-1234":
    if character.isdigit():
        total_digits += int(character)
```

</details>
