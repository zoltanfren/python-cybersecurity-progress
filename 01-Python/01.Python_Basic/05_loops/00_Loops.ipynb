## 1. What Are Loops?

### The problem: repetition without loops

Imagine you want to scan ports 1 through 1024 on a server. Without loops, you would write one line per port — that is 1024 lines for a basic scan:

```python
print("Scanning port 1")
print("Scanning port 2")
print("Scanning port 3")
# ... 1021 more lines ...
print("Scanning port 1024")
```

This is tedious to write, easy to get wrong, and completely unscalable — real scanners check all 65535 ports.

### The solution: loops

A **loop** is an instruction that tells Python: *repeat this block of code*.

The same 1024-line scan becomes two lines:

````python
# Loops: scanning every standard port in two lines
# range(1, 1025) would produce 1, 2, 3 ... 1024
# Shortened to 10 ports here so the output is readable

for port_number in range(1, 11):
    print(f"Scanning port {port_number}")
````

### The DRY principle

In programming there is a rule called **DRY: Don't Repeat Yourself**.

Every time you copy-paste the same code with only a small change, you are probably missing a loop. Loops keep your code short, readable, and easy to maintain — if the logic changes, you update it in one place instead of in a thousand places.

### Python has two types of loops

| Type | Use when |
|---|---|
| `for` | You know how many times to repeat, or you iterate over a sequence |
| `while` | You repeat as long as a condition is true — you do not know in advance how many times |

## 2. The `for` Loop

### Anatomy of a `for` loop

```python
for variable in sequence:
    # indented block — repeats on every iteration
```

- **`variable`** — holds the current value on each iteration. You choose the name.
- **`sequence`** — the thing being iterated over. In this section: `range()`.
- **indented block** — the code that repeats. Everything indented belongs to the loop.

Read it as: *"For each [variable] in [sequence], do this."*

### `range()` — the standard sequence for `for` loops

`range()` generates a sequence of integers. It has three forms:

| Form | Meaning | Example | Produces |
|---|---|---|---|
| `range(stop)` | 0 up to (not including) stop | `range(5)` | 0, 1, 2, 3, 4 |
| `range(start, stop)` | start up to (not including) stop | `range(1, 5)` | 1, 2, 3, 4 |
| `range(start, stop, step)` | start to stop, jumping by step | `range(0, 10, 2)` | 0, 2, 4, 6, 8 |

### ⚠️ stop is exclusive

`range(1, 11)` gives you 1 through **10** — it stops *before* 11. This is the most common beginner mistake with `range()`.

````python
# range(start, stop): scan ports 1 through 5
# range(1, 6) produces 1, 2, 3, 4, 5 — stops before 6

for port_number in range(1, 6):
    print(f"Scanning port {port_number}")
````

````python
# range(start, stop, step): sample every 100th port from 100 to 500
# Useful for quick coverage across a large port range

for port_number in range(100, 501, 100):
    print(f"Sampling port {port_number}")
````

````python
# range(stop) only: simulate 5 generic connection attempts
# range(5) produces 0, 1, 2, 3, 4 — starts at 0 by default

for attempt_number in range(5):
    print(f"Connection attempt {attempt_number}")
````

## 3. Iterating Over Strings

### Strings are sequences too

A `for` loop does not only work with `range()`. It works with any **sequence** — and a string is a sequence of characters.

```python
for character in "hello":
    print(character)
# h
# e
# l
# l
# o
```

### Useful string methods inside a loop

| Method | Returns `True` if the character is... | Example |
|---|---|---|
| `.isdigit()` | a digit 0–9 | `"4".isdigit()` → `True` |
| `.isupper()` | an uppercase letter A–Z | `"A".isupper()` → `True` |
| `.islower()` | a lowercase letter a–z | `"a".islower()` → `True` |
| `.isalpha()` | a letter (any case) | `"x".isalpha()` → `True` |

````python
# String iteration: inspect a password character by character
# Real-world use: first step of a password strength checker

password = "Adm1n!"

for character in password:
    print(f"Character: '{character}'")
````

````python
# Counting digit characters in a password
# digit_count starts at 0 and increases by 1 for every numeric character found

password = "P4ssw0rd!"
digit_count = 0

for character in password:
    if character.isdigit():
        digit_count += 1

print(f"Password '{password}' contains {digit_count} digit(s)")
````

## 4. The `while` Loop

### What is a `while` loop?

A `while` loop repeats **as long as a condition is True**.

```python
while condition:
    # repeats as long as condition is True
    # something inside MUST eventually make condition False!
```

Think of it like a security guard at a door: *"While the alert is active, keep checking."*

### `for` vs `while` — which one to use?

| Situation | Use |
|---|---|
| You know how many times to repeat | `for` |
| You repeat until something changes | `while` |
| Scanning a fixed port range (1 to 1024) | `for` |
| Waiting for the correct password to be entered | `while` |
| Retrying a connection until it succeeds | `while` |

### ⚠️ The infinite loop danger

If the condition never becomes `False`, the loop runs **forever** and your program freezes. Always make sure something inside the loop changes the condition.

````python
# while loop: simulate failed login attempts with lockout
# Loop stops as soon as failed_attempts reaches max_attempts

failed_attempts = 0
max_attempts = 3

while failed_attempts < max_attempts:
    print(f"Failed login attempt {failed_attempts + 1}")
    failed_attempts += 1  # MUST increment — otherwise this loops forever

print(f"Account locked after {max_attempts} failed attempts")
````

````python
# while loop: countdown before an emergency system shutdown
# countdown_seconds decreases each iteration — loop exits when it reaches 0

countdown_seconds = 5

while countdown_seconds > 0:
    print(f"Emergency shutdown in {countdown_seconds} second(s)...")
    countdown_seconds -= 1

print("Shutdown complete")
````

## 5. `break` — Exit a Loop Early

### What is `break`?

`break` immediately stops the current loop and exits it — even if iterations remain.

```python
for port_number in range(1, 1025):
    if port_number == 22:
        break  # stop the entire loop right here
```

### When to use `break`

- You found what you were looking for — no need to keep searching
- A critical condition is met and continuing would be wrong
- The first match is all you need

````python
# break: stop scanning as soon as a critical port is detected
# Port 23 is Telnet — unencrypted, it should never be exposed externally

critical_port = 23

for port_number in range(20, 30):
    print(f"Checking port {port_number}")

    if port_number == critical_port:
        print(f"ALERT: Telnet (port {critical_port}) detected — stopping scan")
        break

print("Scan ended")
````

## 6. `continue` — Skip an Iteration

### What is `continue`?

`continue` skips the rest of the **current** iteration and jumps to the next one. The loop itself keeps running.

```python
for port_number in range(1, 6):
    if port_number == 3:
        continue       # skip port 3, move straight to port 4
    print(port_number)
# prints: 1, 2, 4, 5   (3 is skipped)
```

### `break` vs `continue` — key difference

| Statement | Effect |
|---|---|
| `break` | Exits the **entire** loop |
| `continue` | Skips only the **current** iteration — loop continues |

````python
# continue: skip reserved ports 0 and 1 during a scan
# Ports 0 and 1 are reserved by the OS — attempting to scan them is meaningless

for port_number in range(0, 6):
    if port_number < 2:
        print(f"Skipping reserved port {port_number}")
        continue

    print(f"Scanning port {port_number}")
````

## 7. `pass` — The Placeholder

### What is `pass`?

`pass` does nothing. It is a placeholder used when Python requires a code block but you have nothing to put there yet.

Without `pass`, an empty block causes a `SyntaxError`:

```python
# This crashes — Python expects at least one statement after the colon
for port_number in range(1, 5):

# This works — pass satisfies Python's requirement without doing anything
for port_number in range(1, 5):
    pass
```

### When to use `pass`

- Structuring a script before writing the actual logic
- As a deliberate 'do nothing' in an `if` branch you plan to fill in later
- When building a skeleton program top-down

### `break`, `continue`, `pass` — quick comparison

| Statement | Effect |
|---|---|
| `break` | Exits the loop immediately |
| `continue` | Skips the rest of this iteration |
| `pass` | Does nothing — pure placeholder |

````python
# pass: build a port scan skeleton before writing the logic
# Common practice: structure the flow first, fill in the logic second

for port_number in range(80, 85):
    pass  # TODO: add scan logic here

print("Skeleton loop ran without errors")
````

## 8. Nested Loops

### What are nested loops?

A **nested loop** is a loop inside another loop.

```python
for outer_value in range(...):
    for inner_value in range(...):
        # runs for every combination of outer_value and inner_value
```

For every single iteration of the outer loop, the inner loop runs **completely** from start to finish.

### How to count total iterations

Outer loop runs 3 times, inner loop runs 4 times: **3 × 4 = 12 total iterations** of the inner block.

### Security use case

A real network scanner checks every IP address against every port in scope. That is a nested loop:
- Outer loop: each IP address (or last octet)
- Inner loop: each port to check on that IP

````python
# Nested loops: scan 3 IP last octets across 3 ports
# Outer loop iterates over the last octet of 192.168.1.x (hosts .1, .2, .3)
# Inner loop iterates over ports to check on each host (80, 81, 82)
# Total: 3 hosts x 3 ports = 9 checks

for last_octet in range(1, 4):
    for port_number in range(80, 83):
        print(f"Checking 192.168.1.{last_octet} : port {port_number}")
````

## Summary

| Concept | Syntax | Use it when |
|---|---|---|
| `for` with `range()` | `for x in range(start, stop, step):` | Repeat a known number of times |
| `for` with string | `for char in "text":` | Process each character one by one |
| `while` | `while condition:` | Repeat until a condition changes |
| `break` | `break` inside a loop | Exit the loop immediately |
| `continue` | `continue` inside a loop | Skip to the next iteration |
| `pass` | `pass` inside a block | Placeholder — do nothing |
| Nested loops | `for` inside `for` | Check every combination |

### Key rules to remember

- `range(1, 11)` produces 1 through **10** — stop is always exclusive
- Everything **indented** under the loop belongs to the loop
- A `while` loop with no increment will **run forever**
- `break` exits the loop entirely; `continue` skips only the current iteration; `pass` does nothing
