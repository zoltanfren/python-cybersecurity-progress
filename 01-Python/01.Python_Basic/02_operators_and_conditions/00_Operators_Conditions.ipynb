## 1. Arithmetic Operators

### What Are They?

Arithmetic operators perform calculations — exactly like a calculator. You already know most of them from school.

| Operator | Name | What It Does | Example | Result |
|----------|------|--------------|---------|--------|
| `+` | Addition | Adds two values | `5 + 3` | `8` |
| `-` | Subtraction | Subtracts right from left | `10 - 4` | `6` |
| `*` | Multiplication | Multiplies two values | `3 * 4` | `12` |
| `/` | Division | Divides, always returns float | `10 / 2` | `5.0` |
| `//` | Floor Division | Divides, rounds **down** to integer | `10 // 3` | `3` |
| `%` | Modulus | Returns the **remainder** after division | `10 % 3` | `1` |
| `**` | Exponentiation | Raises to a power | `2 ** 8` | `256` |

### The Modulus Operator `%` — Why It Matters

The modulus is the most unusual one. It returns what is **left over** after division:

```
10 ÷ 3 = 3 with a remainder of 1  →  10 % 3 = 1
20 ÷ 5 = 4 with a remainder of 0  →  20 % 5 = 0
```

**Key uses:**
- Check if a number is even: `number % 2 == 0`
- Check if a number is odd: `number % 2 == 1`
- Wrap values in a range (like a clock)

````python
# Security context: calculate statistics about a scan
total_ports = 65535
scanned_ports = 1024
open_ports = 7

print("=== Scan Statistics ===")
print("Total ports:", total_ports)
print("Scanned:", scanned_ports)
print("Remaining:", total_ports - scanned_ports)      # subtraction
print("Open ports:", open_ports)
print("Closed ports:", scanned_ports - open_ports)    # subtraction
print("Open %:", (open_ports / scanned_ports) * 100)  # division + multiplication
print()

# Floor division: how many full batches of 256 ports fit in 1024?
batch_size = 256
full_batches = scanned_ports // batch_size
remaining = scanned_ports % batch_size
print(f"Batches of {batch_size}: {full_batches} complete, {remaining} remaining")
````

````python
# Modulus examples — the remainder operator
print("10 % 3 =", 10 % 3)   # 1 — because 3×3=9, leftover is 1
print("20 % 5 =", 20 % 5)   # 0 — divides exactly, no remainder
print("17 % 4 =", 17 % 4)   # 1 — because 4×4=16, leftover is 1
print()

# Practical use: check even/odd
port = 22
if port % 2 == 0:
    print(f"Port {port} is even")
else:
    print(f"Port {port} is odd")
````

### Assignment Operators — Updating Variables in Place

These are shortcuts for updating a variable using itself:

| Shorthand | Equivalent | Meaning |
|-----------|------------|--------|
| `x += 3` | `x = x + 3` | Add 3 to x |
| `x -= 3` | `x = x - 3` | Subtract 3 from x |
| `x *= 3` | `x = x * 3` | Multiply x by 3 |
| `x //= 2` | `x = x // 2` | Floor-divide x by 2 |

````python
# Counting failed login attempts — a real security pattern
failed_attempts = 0
print("Start:", failed_attempts)

failed_attempts += 1  # Same as: failed_attempts = failed_attempts + 1
print("After 1st fail:", failed_attempts)

failed_attempts += 1
print("After 2nd fail:", failed_attempts)

failed_attempts += 1
print("After 3rd fail:", failed_attempts)

if failed_attempts >= 3:
    print("⚠️  Account locked — too many failed attempts")
````

## 2. Comparison Operators

### What Are They?

Comparison operators compare two values and always return **`True` or `False`**. They are the foundation of every decision your code makes.

| Operator | Name | Question It Answers | Example | Result |
|----------|------|---------------------|---------|--------|
| `==` | Equal to | Are they the same value? | `5 == 5` | `True` |
| `!=` | Not equal to | Are they different? | `5 != 3` | `True` |
| `>` | Greater than | Is left strictly bigger? | `10 > 5` | `True` |
| `<` | Less than | Is left strictly smaller? | `3 < 7` | `True` |
| `>=` | Greater or equal | Is left bigger or the same? | `5 >= 5` | `True` |
| `<=` | Less or equal | Is left smaller or the same? | `4 <= 6` | `True` |

### ⚠️ The Most Common Beginner Mistake: `=` vs `==`

```python
port = 80       # ASSIGNMENT — stores 80 in port
port == 80      # COMPARISON — asks "is port equal to 80?" → True
```

Using `=` inside an `if` condition is a `SyntaxError` in Python. Good news: Python will tell you!

````python
# Comparing port numbers
port = 443

print("port == 443:", port == 443)   # True  — exact match
print("port == 80:",  port == 80)    # False — different value
print("port != 80:",  port != 80)    # True  — they are different
print("port > 1024:", port > 1024)   # False — 443 is below 1024
print("port < 1024:", port < 1024)   # True  — well-known port range
print("port >= 443:", port >= 443)   # True  — equal counts as >=
print("port <= 443:", port <= 443)   # True  — equal counts as <=
````

### Chaining Comparisons — A Python Superpower

Python lets you chain comparisons in a way that reads like plain English:

```python
# Traditional (other languages)
if port > 1024 and port <= 49151:

# Pythonic — reads like math notation
if 1024 < port <= 49151:
```

This checks all conditions left to right. If any one fails, the result is `False`.

````python
# Port range classification using chained comparisons
port = 8080

if 1 <= port <= 1023:
    print(f"Port {port}: Well-known (system) port")
elif 1024 <= port <= 49151:
    print(f"Port {port}: Registered port")
elif 49152 <= port <= 65535:
    print(f"Port {port}: Dynamic/private port")
else:
    print(f"Port {port}: Invalid port number")
````

### Comparing Strings — Lexicographic Order

Python can compare strings too. It uses **alphabetical (lexicographic) order** — the same order as a dictionary.

````python
# String comparison
print("'admin' == 'admin':", 'admin' == 'admin')    # True
print("'admin' == 'Admin':", 'admin' == 'Admin')    # False — case sensitive!
print("'apple' < 'banana':", 'apple' < 'banana')   # True — 'a' comes before 'b'
print()

# Security relevance: passwords are case-sensitive
entered_password = "Secret123"
correct_password = "secret123"

if entered_password == correct_password:
    print("Access granted")
else:
    print("Access denied — passwords are case-sensitive!")
````

## 3. Conditions — Making Decisions

### The Concept

Conditions let your program **choose** what to do, like you do every day:

```
IF the door is locked:
    Use your key
ELSE:
    Just walk in
```

Python uses `if`, `elif`, and `else` to express this logic.

### The `if` Statement

```python
if condition:
    # code runs only when condition is True
    # indented 4 spaces — this marks it as "inside" the if
```

Python evaluates the condition. If it is `True`, the indented block runs. If it is `False`, Python skips it entirely.

**The colon `:` and the indentation are mandatory.** Missing either one is a `SyntaxError`.

````python
# Basic if — only runs when condition is True
cvss_score = 9.1

if cvss_score >= 9.0:
    print("CRITICAL vulnerability detected")  # inside the if (4 spaces)
    print("Immediate patching required")       # also inside the if

print("Scan complete")  # outside the if — always runs
````

````python
# When condition is False — the block is skipped
cvss_score = 5.0

if cvss_score >= 9.0:
    print("CRITICAL")  # skipped — 5.0 is not >= 9.0

print("Scan complete")  # always runs
````

### The `else` Statement

`else` defines what happens when the `if` condition is `False`. It is the fallback.

```python
if condition:
    # runs when True
else:
    # runs when False — the fallback
```

**Important:** Only one branch ever runs — either the `if` OR the `else`, never both.

````python
# Login check — if/else
username = "alice"
password = "correct_password"

if username == "alice" and password == "correct_password":
    print("✅ Access granted — welcome, alice")
else:
    print("❌ Access denied — invalid credentials")
````

### The `elif` Statement

`elif` means "else if" — it checks another condition if the previous one was `False`.

You can have as many `elif` blocks as you need.

```python
if condition_1:
    # runs if condition_1 is True
elif condition_2:
    # runs if condition_1 was False AND condition_2 is True
elif condition_3:
    # runs if 1 and 2 were False AND condition_3 is True
else:
    # runs if ALL conditions above were False
```

### Critical Rule: First Match Wins

Python evaluates from **top to bottom** and stops at the **first True condition**. Everything after is skipped.

````python
# CVSS severity classification — a real-world example
cvss_score = 7.5

if cvss_score >= 9.0:
    severity = "CRITICAL"   # 9.0 - 10.0
elif cvss_score >= 7.0:
    severity = "HIGH"       # 7.0 - 8.9
elif cvss_score >= 4.0:
    severity = "MEDIUM"     # 4.0 - 6.9
elif cvss_score > 0:
    severity = "LOW"        # 0.1 - 3.9
else:
    severity = "NONE"       # 0.0

print(f"CVSS {cvss_score} → {severity}")
````

````python
# First-match behavior — why ORDER matters
score = 95

if score >= 90:
    print("Grade A")   # This runs — score 95 matches
elif score >= 80:       # True, but Python NEVER checks this — already matched above
    print("Grade B")
elif score >= 70:
    print("Grade C")

print()

# WRONG order — produces incorrect result
cvss_score = 9.5
if cvss_score >= 4.0:       # This matches first! Wrong classification
    print("MEDIUM (wrong — order matters!)")
elif cvss_score >= 7.0:
    print("HIGH")
elif cvss_score >= 9.0:
    print("CRITICAL")
````

### Nested Conditions

You can put `if` statements inside other `if` statements. Each level adds 4 more spaces of indentation.

````python
# Nested if — firewall rule check
port = 22
source_ip = "192.168.1.100"
is_internal = True

if port == 22:                          # Level 1: is it SSH?
    if is_internal:                     # Level 2: is it internal?
        print("✅ SSH from internal — allowed")
    else:
        print("⚠️  SSH from external — blocked")
else:
    print(f"Port {port}: standard policy applies")
````

## 4. Logical Operators — Combining Conditions

### What Are They?

Logical operators let you combine multiple conditions into one. In real security decisions, you rarely check just one thing:

- "Block this IP **if** it's in the blocklist **AND** the port is 22"
- "Alert **if** the login failed **OR** the session is expired"
- "Allow **if** the user is **NOT** anonymous"

Python has three logical operators: `and`, `or`, `not`.

### The `and` Operator

`and` returns `True` **only when both conditions are True**.

**Truth table:**

| Left | Right | Result |
|------|-------|--------|
| `True` | `True` | ✅ `True` |
| `True` | `False` | ❌ `False` |
| `False` | `True` | ❌ `False` |
| `False` | `False` | ❌ `False` |

````python
# Authentication: both username AND password must match
username = "alice"
password = "secure_pass_123"

if username == "alice" and password == "secure_pass_123":
    print("✅ Login successful")
else:
    print("❌ Login failed")
````

````python
# All four combinations of and
print("True and True:",   True and True)    # True
print("True and False:",  True and False)   # False
print("False and True:",  False and True)   # False
print("False and False:", False and False)  # False
````

### The `or` Operator

`or` returns `True` **when at least one condition is True**.

**Truth table:**

| Left | Right | Result |
|------|-------|--------|
| `True` | `True` | ✅ `True` |
| `True` | `False` | ✅ `True` |
| `False` | `True` | ✅ `True` |
| `False` | `False` | ❌ `False` |

````python
# Alert if firewall OR antivirus is off
firewall_active  = True
antivirus_active = False

if not firewall_active or not antivirus_active:
    print("⚠️  Security gap detected — check your defenses")
else:
    print("✅ All defenses active")
````

````python
# All four combinations of or
print("True or True:",   True or True)    # True
print("True or False:",  True or False)   # True
print("False or True:",  False or True)   # True
print("False or False:", False or False)  # False
````

### The `not` Operator

`not` **reverses** a boolean value. `True` becomes `False`, and `False` becomes `True`.

| Input | Result |
|-------|--------|
| `not True` | `False` |
| `not False` | `True` |

````python
# not — reverses the value
is_authenticated = False

if not is_authenticated:
    print("Access denied — please log in first")

print("not True =",  not True)   # False
print("not False =", not False)  # True
````

### Short-Circuit Evaluation

Python is **lazy** — it stops evaluating as soon as the result is certain:

- `and`: if the **left side is False**, the right side is **never evaluated** (result is already False)
- `or`: if the **left side is True**, the right side is **never evaluated** (result is already True)


1. **Performance** — avoids unnecessary computation
2. **Safety** — avoids errors in the second condition
3. **Security** — check cheap/safe condition first

````python
# Safe pattern: check list is not empty BEFORE accessing elements
# Without short-circuit, packets[0] would crash if packets is empty

packets = ["SYN", "ACK", "RST"]

# packets is truthy (non-empty) → Python evaluates the right side
if packets and packets[0] == "SYN":
    print("SYN packet found — possible scan")

# Empty list — Python stops at 'packets' (falsy), never checks packets[0]
empty_packets = []
if empty_packets and empty_packets[0] == "SYN":
    print("This won't crash because short-circuit stops here")
else:
    print("No packets to analyze")
````

### Operator Precedence with Logical Operators

When you mix `and`, `or`, `not` in one expression, Python evaluates in this order:

1. `not` — highest priority
2. `and`
3. `or` — lowest priority

**Best practice: use parentheses** to make your intent explicit.

````python
# Ambiguous without parentheses
firewall = True
ids = False
patched = True

# Python reads this as: firewall and (ids or patched)
result_a = firewall and ids or patched
print("Without parens:", result_a)

# With parentheses — intention is crystal clear
result_b = (firewall and ids) or patched
result_c = firewall and (ids or patched)
print("(firewall and ids) or patched:", result_b)
print("firewall and (ids or patched):", result_c)
````

## 5. The `in` Operator — Membership Testing

### What Is It?

The `in` operator asks: **"Is this value inside this collection?"**

It returns `True` or `False` and works on lists, tuples, sets, strings, and dictionaries.

```python
item in collection        # True if item exists in collection
item not in collection    # True if item does NOT exist
```

````python
# Port blocklist check
dangerous_ports = [21, 23, 135, 445, 3389]

for port in [22, 23, 80, 3389]:
    if port in dangerous_ports:
        print(f"Port {port}: ⚠️  DANGEROUS")
    else:
        print(f"Port {port}: ✅ acceptable")
````

````python
# IP blocklist check
blocked_ips = ["192.168.1.100", "10.0.0.50", "172.16.0.25"]
incoming_ip = "192.168.1.100"

if incoming_ip in blocked_ips:
    print(f"🚫 {incoming_ip} is blocked — connection refused")
else:
    print(f"✅ {incoming_ip} allowed")

# not in — checking absence
allowed_users = ["alice", "bob", "charlie"]
user = "eve"

if user not in allowed_users:
    print(f"❌ User '{user}' is not authorized")
````

````python
# in works on strings too — check for keywords in a log
log_entry = "FAILED login attempt from 192.168.1.100 at 10:45"

if "FAILED" in log_entry:
    print("⚠️  Failed login detected in logs")

if "192.168" in log_entry:
    print("📍 Internal IP detected")
````

## 6. `is` vs `==` — Identity vs Equality

### The Difference

| Operator | Question | Checks |
|----------|----------|--------|
| `==` | Are the **values** the same? | Content |
| `is` | Are they the **same object in memory**? | Identity |

Two variables can hold **equal values** but point to **different objects** in memory:

```
x = [1, 2, 3]   ──→  Object A in memory
y = [1, 2, 3]   ──→  Object B in memory (different!)

x == y  →  True   (same values)
x is y  →  False  (different objects)
```

````python
# == vs is with lists
list_a = [1, 2, 3]
list_b = [1, 2, 3]  # Same content, different object
list_c = list_a     # Same object (both point to the same place)

print("list_a == list_b:", list_a == list_b)  # True  — same values
print("list_a is list_b:", list_a is list_b)  # False — different objects
print()
print("list_a == list_c:", list_a == list_c)  # True
print("list_a is list_c:", list_a is list_c)  # True  — same object!
````

### When to Use Each

**Use `==`** for comparing values — almost always what you want.

**Use `is`** only for:
- Comparing to `None` → **always use `is None`**, never `== None`
- Comparing to `True` / `False` singletons (rare)

````python
# The canonical use of 'is': checking for None
scan_result = None  # No result yet

if scan_result is None:      # ✅ Correct — use 'is' with None
    print("No scan result available yet")

# if scan_result == None:    # ⚠️  Works but not recommended by PEP 8

# Value comparison — always use ==
status = "open"
if status == "open":         # ✅ Correct — value comparison
    print("Port is open")
````

## 7. Truthy and Falsy Values

### The Concept

In Python, **every value** can be used in a boolean context (inside an `if`). Not just `True` and `False`.

**Falsy** — treated as `False` in an `if`:
- `False`, `None`
- `0`, `0.0` (any zero)
- `""` (empty string)
- `[]`, `()`, `{}`, `set()` (any empty collection)

**Truthy** — everything else is treated as `True`:
- Any non-zero number
- Any non-empty string
- Any non-empty collection

````python
# See which values are truthy/falsy
test_values = [0, 1, -1, "", "text", [], [0], None, False, True, {}, {"a": 1}]

for val in test_values:
    label = "truthy" if val else "falsy "
    print(f"  {repr(val):15} is {label}")
````

````python
# Practical security use: check if collections are empty
open_ports   = [22, 80]     # non-empty
blocked_ips  = []           # empty
scan_result  = None

# Pythonic way — read as plain English
if open_ports:
    print(f"Found {len(open_ports)} open ports")

if not blocked_ips:
    print("Blocklist is empty — consider adding entries")

if scan_result is None:
    print("No scan result yet")
````

## 8. Ternary Expression — One-Line Conditions

### What Is It?

Python has a compact syntax for simple `if/else` assignments:

```python
value = result_if_true if condition else result_if_false
```

**When to use it:** Simple assignments where both branches are one value. Avoid it for complex logic — it reduces readability.

````python
# Traditional if/else
cvss = 9.0
if cvss >= 9.0:
    label = "CRITICAL"
else:
    label = "not critical"
print("Traditional:", label)

# Ternary — same result, one line
label = "CRITICAL" if cvss >= 9.0 else "not critical"
print("Ternary:", label)
````

````python
# Useful inline in print statements and f-strings
port = 22
is_encrypted = port in [22, 443, 993, 995]

print(f"Port {port} is {'encrypted' if is_encrypted else 'unencrypted'}")

# Classify multiple ports
for p in [22, 80, 443, 21]:
    status = "safe" if p in [22, 443] else "review"
    print(f"  Port {p}: {status}")
````

## 9. Operator Precedence — The Order Python Evaluates Things

### Why This Matters

When you write `2 + 3 * 4`, do you get `20` or `14`? Python evaluates `*` before `+`, so the answer is `14`.

The same applies to all operators. Python follows a strict order:

| Priority | Operator(s) | Category |
|----------|------------|----------|
| 1 (highest) | `**` | Exponentiation |
| 2 | `*`, `/`, `//`, `%` | Multiplication / Division |
| 3 | `+`, `-` | Addition / Subtraction |
| 4 | `<`, `>`, `<=`, `>=`, `==`, `!=`, `in`, `is` | Comparisons |
| 5 | `not` | Logical NOT |
| 6 | `and` | Logical AND |
| 7 (lowest) | `or` | Logical OR |

**Golden rule: when in doubt, use parentheses `()`.**

````python
# Arithmetic precedence
print("2 + 3 * 4 =", 2 + 3 * 4)        # 14 — * before +
print("(2 + 3) * 4 =", (2 + 3) * 4)    # 20 — () overrides
print("2 ** 3 + 1 =", 2 ** 3 + 1)      # 9  — ** before +
print()
````

````python
# Logical precedence
# not > and > or

a, b, c = True, False, True

# Python reads: (a and b) or c
result1 = a and b or c
print("a and b or c:        ", result1)  # True

# With parentheses — explicit
result2 = a and (b or c)
print("a and (b or c):      ", result2)  # True

result3 = (a and b) or c
print("(a and b) or c:      ", result3)  # True

# not has highest priority
print("not a and b:         ", not a and b)   # False — reads as (not a) and b
print("not (a and b):       ", not (a and b)) # True  — explicit grouping
````

## 10. Best Practices — Writing Clean Conditions

Following PEP 8 and these conventions will make your code easier to read and review:

```python
# ✅ Use 'is None', never '== None'
if value is None:

# ✅ Use implicit truthiness for collections
if open_ports:         # instead of: if len(open_ports) > 0

# ✅ Use parentheses for complex conditions
if (port in dangerous) and (not is_internal):

# ✅ Avoid Yoda conditions (value on the left, variable on right)
if score >= 9.0:       # instead of: if 9.0 <= score

# ✅ Put most likely condition first in 'or' (short-circuit wins)
if firewall_on or vpn_on:   # check the common case first
```

````python
# Putting it all together — a realistic security check
def check_connection(ip, port, failed_attempts, is_internal):
    """Evaluate whether a connection attempt should be allowed."""
    blocked_ips    = ["10.0.0.50", "192.168.1.200"]
    dangerous_ports = [21, 23, 135, 445]
    MAX_ATTEMPTS   = 3

    # Each check is clear and named
    is_blocked   = ip in blocked_ips
    is_dangerous = port in dangerous_ports
    is_lockout   = failed_attempts >= MAX_ATTEMPTS

    # Final decision
    if is_blocked or is_lockout:
        return "❌ DENIED — blocked or locked out"
    elif is_dangerous and not is_internal:
        return "⚠️  DENIED — dangerous port from external source"
    else:
        return "✅ ALLOWED"

# Test cases
print(check_connection("10.0.0.50",   22,  0, False))  # blocked IP
print(check_connection("1.2.3.4",     23,  0, False))  # dangerous port
print(check_connection("192.168.1.5", 23,  0, True))   # dangerous but internal
print(check_connection("1.2.3.4",     80,  0, False))  # normal request
print(check_connection("1.2.3.4",     80,  5, False))  # locked out
````

## 📝 Summary

| Concept | Key Points |
|---------|------------|
| **Arithmetic operators** | `+` `-` `*` `/` `//` `%` `**` — `%` gives the remainder, `//` gives the integer quotient |
| **Assignment shortcuts** | `x += 1` is shorthand for `x = x + 1` |
| **Comparison operators** | `==` `!=` `>` `<` `>=` `<=` — always return `True` or `False` |
| **`=` vs `==`** | `=` assigns a value, `==` compares values — never confuse them |
| **Chaining** | `1 < x < 10` is valid Python — reads left to right |
| **`if/elif/else`** | First match wins — order matters, only one branch runs |
| **`and`** | Both must be True |
| **`or`** | At least one must be True |
| **`not`** | Reverses True/False |
| **Short-circuit** | `and` stops at first False, `or` stops at first True |
| **`in` / `not in`** | Membership test — works on lists, sets, strings, dicts |
| **`is` vs `==`** | `is` checks identity (same object), `==` checks value — use `is` only for `None` |
| **Truthy/Falsy** | `0`, `""`, `[]`, `None` are falsy — everything else is truthy |
| **Ternary** | `x if condition else y` — compact but use only for simple cases |
| **Precedence** | `**` → `* / // %` → `+ -` → comparisons → `not` → `and` → `or` |
