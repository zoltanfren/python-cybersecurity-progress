**Exercise 14 — Port classifier**  

Create a variable `target_port` with value `443`.  
Create a variable `port_category` using `if/elif/else` with **chained comparisons**:
- Between `1` and `1023` (inclusive) → `"well-known"`
- Between `1024` and `49151` (inclusive) → `"registered"`
- Between `49152` and `65535` (inclusive) → `"dynamic"`
- Otherwise → `"invalid"`

````python
# Exercise 14
target_port = 443
port_category = "invalid"
if 1 <= target_port <= 1023: port_category = "well-known" 
elif 1024 <= target_port <= 49151: port_category = "registered" 
elif 49152 <= target_port <= 65535: port_category = "dynamic" 
````

    port_category = "well-known"

Create a variable `alert_level` using `if/elif/else`:

- `"GREEN"` otherwise

Create a variable `notify_admin` that is `True` if `alert_level` is `"RED"` or `"ORANGE"`, **and** `admin_online` is `True`.
