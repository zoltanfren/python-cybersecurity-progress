# Module 5 - Imports — Using Code That Already Exists

### The Problem `import` Solves

Python comes with thousands of useful tools. But if **all of them were loaded every time you ran Python**, startup would be slow and memory usage enormous.

Instead, Python splits its tools into two categories:
- **Built-in functions** — always available, no setup needed
- **Modules** — optional toolboxes you load on demand with `import`

---

### Built-in Functions — Always Ready

Built-in functions are **baked into Python itself**. You can use them anywhere, at any time, with no `import`. They are the most fundamental tools.

**Common built-in functions:**

| Function | What it does | Example |
|----------|-------------|----------|
| `print()` | Display output | `print("Hello")` |
| `input()` | Read user input | `input("Name: ")` |
| `len()` | Count items | `len([1, 2, 3])` → 3 |
| `int()` | Convert to integer | `int("443")` → 443 |
| `str()` | Convert to string | `str(443)` → `"443"` |
| `type()` | Get the type | `type(443)` → `<class 'int'>` |
| `range()` | Generate a number sequence | `range(0, 5)` |

These require **zero setup** — they are always there.

````python
# Built-in functions need NO import — they just work
ports = [443, 22, 80, 3306, 8080]

print("List:", ports)               # print() — built-in
print("Length:", len(ports))        # len()   — built-in
print("Type:", type(ports))         # type()  — built-in
print("As string:", str(ports[0]))  # str()   — built-in
````

---
## Modules — Toolboxes You Load On Demand

A **module** is a file containing Python code — functions, variables, constants — that you can load into your script. Python ships with over 200 modules in its **Standard Library**, and thousands more exist online.

**Why isn't everything built-in?** Think of a toolbox analogy:
- Your **hands** = built-in functions (always with you, no effort)
- A **toolbox** = a module (you bring it only when you need it)

You wouldn't carry a full toolbox everywhere just in case. Same logic applies in Python — loading every module at startup would waste time and memory.

### The Four Import Syntaxes

```python
# Method 1: Import the whole module
import random
random.randint(1, 10)      # Must prefix with module name

# Method 2: Import a specific function from a module
from random import randint
randint(1, 10)             # No prefix needed

# Method 3: Import with an alias
import random as rnd
rnd.randint(1, 10)         # Use alias as prefix

# Method 4: Import everything (NOT recommended)
from random import *       # Pollutes your namespace — avoid this
```

**Best practice:** Use Method 1 or Method 2.  
**Where to put imports?** Always at the **very top** of the file, before any other code.

### Concrete Example: Generating a Random Number

There is **no built-in function** for randomness — you need the `random` module. Watch what happens without importing first:

````python
# Without import — this crashes with NameError
number = randint(1, 100)  # ← NameError: name 'randint' is not defined

print("randint doesn't exist yet — Python has no idea what it is")
````

````python
# Method 1: import the whole module
import random

# Access functions with module_name.function_name()
# The prefix tells you exactly where the function comes from
number = random





int("Random number:", number)
````

````python
# Method 2: import only what you need
from random import randint

# randint is now directly in your namespace — no prefix needed
number = randint(1, 100)
print("Random number:", number)
````

````python
# Other useful functions inside the random module
import random

# Random integer (both ends inclusive)
port = random.randint(1024, 65535)
print(f"Random registered port: {port}")

# Random float between 0.0 and 1.0
risk_score = random.random()
print(f"Risk score: {risk_score:.2f}")

# Pick one random item from a list
targets = ["192.168.1.1", "192.168.1.2", "192.168.1.3"]
chosen = random.choice(targets)
print(f"Random target: {chosen}")

# Shuffle a list in place
ports = [22, 80, 443, 3306, 8080]
random.shuffle(ports)
print(f"Shuffled ports: {ports}")
````

### Built-in vs Imported — The Full Picture

| | Built-in | Standard Library | Third-party |
|--|----------|-----------------|-------------|
| **Needs `import`?** | ❌ No | ✅ Yes | ✅ Yes |
| **Needs `pip install`?** | ❌ No | ❌ No | ✅ Yes |
| **Always available?** | ✅ Yes | ✅ If Python installed | ❌ Must install first |
| **Examples** | `print()` `len()` `int()` | `random` `os` `hashlib` | `requests` `scapy` `nmap` |
