### Exercise 7

Create a function called `count_high_ports(ports)` that:
- `ports` is a list of integers
- **Returns** the count of ports that are `>= 1024`

Expected: `count_high_ports([22, 80, 443, 8080, 3306])` → `2` (8080 and 3306)

````python
# Exercise 7
````

<details>
<summary><span style="color:#e74c3c; font-size:1.1em;"><strong>💡 SOLUTION</strong></span></summary>

```python
# SOLUTION
def count_high_ports(ports):
    count = 0
    for port in ports:
        if port >= 1024:
            count += 1
    return count
```

### Exercise 8

Create a function called `filter_by_severity(vulnerabilities, min_severity)` that:
- `vulnerabilities` is a list of dicts, each with keys `"name"` and `"cvss"`
- **Returns** a new list containing only vulnerabilities where `cvss >= min_severity`

Example input:
```python
[
    {"name": "SQLi", "cvss": 9.8},
    {"name": "XSS", "cvss": 6.1},
    {"name": "CSRF", "cvss": 8.8}
]
```

Expected: `filter_by_severity(vulns, 8.0)` → `[{"name": "SQLi", "cvss": 9.8}, {"name": "CSRF", "cvss": 8.8}]`

````python
# Exercise 8
````

<details>
<summary><span style="color:#e74c3c; font-size:1.1em;"><strong>💡 SOLUTION</strong></span></summary>

```python
# SOLUTION
def filter_by_severity(vulnerabilities, min_severity):
    result = []
    for vuln in vulnerabilities:
        if vuln["cvss"] >= min_severity:
            result.append(vuln)
    return result
```

### Exercise 13

Create a function called `calculate_average_score(*scores)` that:

### Exercise 15

Create a function called `merge_scan_results(base_result, **additional)` that:
- `base_result` is a dict (the base scan result)
- `**additional` accepts any number of keyword arguments
- **Returns** a new dict that merges `base_result` with the additional keyword arguments

Example:
```python
merge_scan_results(
    {"ip": "192.168.1.10", "port": 80},
    status="open",
    service="HTTP"
)
```
Expected result: `{"ip": "192.168.1.10", "port": 80, "status": "open", "service": "HTTP"}`

````python
# Exercise 15
````

<details>
<summary><span style="color:#e74c3c; font-size:1.1em;"><strong>💡 SOLUTION</strong></span></summary>

```python
# SOLUTION
def merge_scan_results(base_result, **additional):
    # Create a copy of base_result to avoid modifying the original
    merged = base_result.copy()
    # Add all additional keyword arguments
    for key, value in additional.items():
        merged[key] = value
    return merged
```

</details>

## Testing

Run the cell below to check all 15 exercises at once.
Read each error message carefully — it tells you exactly what was expected and what you produced.

Do not modify this cell.

````python
import unittest


class TestFunctionsDrills(unittest.TestCase):

    # Part 1 — Basic Functions

    def test_01_greet(self):
        self.assertTrue(callable(greet),
            "greet must be a callable function")

    def test_02_double(self):
        self.assertEqual(double(5), 10,
            "double(5) should return 10")
        self.assertEqual(double(0), 0,
            "double(0) should return 0")
        self.assertEqual(double(-3), -6,
            "double(-3) should return -6")

    def test_03_is_port_open(self):
        self.assertIsInstance(is_port_open(80, [80, 443]), bool,
            "is_port_open must return a bool, not a string or int")
        self.assertEqual(is_port_open(80, [80, 443, 22]), True,
            "is_port_open(80, [80, 443, 22]) should return True — 80 is in the list")
        self.assertEqual(is_port_open(8080, [80, 443, 22]), False,
            "is_port_open(8080, [80, 443, 22]) should return False — 8080 is not in the list")

    # Part 2 — Parameters and Return Values

    def test_04_calculate_risk(self):
        self.assertIsInstance(calculate_risk(5, 3), int,
            "calculate_risk should return an int when both inputs are ints")
        self.assertEqual(calculate_risk(5, 3), 15,
            "calculate_risk(5, 3) should return 15 (5 * 3)")
        self.assertEqual(calculate_risk(8, 7), 56,
            "calculate_risk(8, 7) should return 56 (8 * 7)")

    def test_05_create_url(self):
        self.assertIsInstance(create_url("example.com"), str,
            "create_url must return a string")
        self.assertEqual(create_url("example.com"), "https://example.com",
            "create_url('example.com') should use default protocol 'https'")
        self.assertEqual(create_url("example.com", "http"), "http://example.com",
            "create_url('example.com', 'http') should override protocol to 'http'")

    def test_06_format_ip_port(self):
        self.assertIsInstance(format_ip_port("192.168.1.10", 80), str,
            "format_ip_port must return a string")
        self.assertEqual(format_ip_port("192.168.1.10", 80), "192.168.1.10:80",
            "format_ip_port should format as 'ip:port'")
        self.assertEqual(format_ip_port("10.0.0.1", 443), "10.0.0.1:443",
            "format_ip_port('10.0.0.1', 443) should return '10.0.0.1:443'")

    # Part 3 — Working with Lists and Dicts

    def test_07_count_high_ports(self):
        self.assertIsInstance(count_high_ports([22, 80]), int,
            "count_high_ports must return an int")
        self.assertEqual(count_high_ports([22, 80, 443, 8080, 3306]), 2,
            "count_high_ports([22, 80, 443, 8080, 3306]) should return 2 — ports 8080 and 3306 are >= 1024")
        self.assertEqual(count_high_ports([22, 80, 443]), 0,
            "count_high_ports([22, 80, 443]) should return 0 — all ports are < 1024")

    def test_08_filter_by_severity(self):
        vulns = [
            {"name": "SQLi", "cvss": 9.8},
            {"name": "XSS", "cvss": 6.1},
            {"name": "CSRF", "cvss": 8.8}
        ]
        result = filter_by_severity(vulns, 8.0)
        self.assertIsInstance(result, list,
            "filter_by_severity must return a list")
        self.assertEqual(len(result), 2,
            "filter_by_severity(vulns, 8.0) should return 2 vulnerabilities (SQLi and CSRF)")
        self.assertEqual(result[0]["name"], "SQLi",
            "First result should be SQLi (cvss 9.8 >= 8.0)")
        self.assertEqual(result[1]["name"], "CSRF",
            "Second result should be CSRF (cvss 8.8 >= 8.0)")

    def test_09_get_service_name(self):
        port_map = {22: "SSH", 80: "HTTP", 443: "HTTPS"}
        self.assertIsInstance(get_service_name(80, port_map), str,
            "get_service_name must return a string")
        self.assertEqual(get_service_name(80, port_map), "HTTP",
            "get_service_name(80, ...) should return 'HTTP'")
        self.assertEqual(get_service_name(3306, port_map), "Unknown",
            "get_service_name(3306, ...) should return 'Unknown' when port not in dict")

    # Part 4 — Advanced Techniques

    def test_10_sum_all(self):
        self.assertIsInstance(sum_all(1, 2, 3), int,
            "sum_all must return an int when all args are ints")
        self.assertEqual(sum_all(1, 2, 3), 6,
            "sum_all(1, 2, 3) should return 6")
        self.assertEqual(sum_all(10), 10,
            "sum_all(10) should return 10")
        self.assertEqual(sum_all(1, 2, 3, 4, 5), 15,
            "sum_all(1, 2, 3, 4, 5) should return 15")

    def test_11_is_high_port(self):
        self.assertTrue(callable(is_high_port),
            "is_high_port must be callable (a lambda function)")
        self.assertIsInstance(is_high_port(8080), bool,
            "is_high_port must return a bool")
        self.assertEqual(is_high_port(8080), True,
            "is_high_port(8080) should return True (8080 >= 1024)")
        self.assertEqual(is_high_port(80), False,
            "is_high_port(80) should return False (80 < 1024)")

    def test_12_build_config(self):
        result = build_config(hostname="web01", port=443)
        self.assertIsInstance(result, dict,
            "build_config must return a dict")
        self.assertEqual(result, {"hostname": "web01", "port": 443},
            "build_config(hostname='web01', port=443) should return {'hostname': 'web01', 'port': 443}")

    # Part 5 — Putting It Together

    def test_13_calculate_average_score(self):
        avg = calculate_average_score(9.8, 7.5, 8.8)
        self.assertIsInstance(avg, float,
            "calculate_average_score should return a float when given floats")
        self.assertAlmostEqual(avg, 8.7, places=5,
            msg="calculate_average_score(9.8, 7.5, 8.8) should return 8.7")
        self.assertEqual(calculate_average_score(), 0,
            "calculate_average_score() with no arguments should return 0")

    def test_14_is_privileged_port(self):
        self.assertIsNotNone(is_privileged_port.__doc__,
            "is_privileged_port must have a docstring")
        self.assertGreater(len(is_privileged_port.__doc__.strip()), 0,
            "is_privileged_port docstring must not be empty")
        self.assertIsInstance(is_privileged_port(80), bool,
            "is_privileged_port must return a bool")
        self.assertEqual(is_privileged_port(80), True,
            "is_privileged_port(80) should return True (80 < 1024)")
        self.assertEqual(is_privileged_port(8080), False,
            "is_privileged_port(8080) should return False (8080 >= 1024)")

    def test_15_merge_scan_results(self):
        base = {"ip": "192.168.1.10", "port": 80}
        result = merge_scan_results(base, status="open", service="HTTP")
        self.assertIsInstance(result, dict,
            "merge_scan_results must return a dict")
        expected = {"ip": "192.168.1.10", "port": 80, "status": "open", "service": "HTTP"}
        self.assertEqual(result, expected,
            "merge_scan_results should merge base_result with additional kwargs")
        # Ensure original dict was not modified
        self.assertEqual(base, {"ip": "192.168.1.10", "port": 80},
            "merge_scan_results should not modify the original base_result dict")


unittest.main(argv=[""], verbosity=2, exit=False)
````
