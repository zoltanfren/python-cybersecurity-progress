## 1. What Are Functions?

### The problem: repetition without functions

Imagine you need to check if port 22 is open on three different servers. Without functions:

```python
# Server 1
print("Connecting to 192.168.1.10...")
if 22 in [22, 80, 443]:
    print("SSH is open on 192.168.1.10")

# Server 2
print("Connecting to 192.168.1.11...")
if 22 in [80, 443]:
    print("SSH is closed on 192.168.1.11")

# Server 3...
```

This violates the DRY principle (Don't Repeat Yourself) that you learned in the Loops module.

### The solution: functions

A **function** is a named block of code that you write once and can execute many times.

```python
def check_ssh(ip_address, open_ports):
    print(f"Connecting to {ip_address}...")
    if 22 in open_ports:
        print(f"SSH is open on {ip_address}")
    else:
        print(f"SSH is closed on {ip_address}")

# Use it on all servers
check_ssh("192.168.1.10", [22, 80, 443])
check_ssh("192.168.1.11", [80, 443])
check_ssh("192.168.1.12", [22, 3306])
```


### The five benefits of functions

| Benefit | Meaning |
|---|---|
| **Reusability** | Write once, use many times |
| **Organization** | Break complex code into small pieces |
| **Maintainability** | Fix bugs in one place |
| **Testability** | Easy to test isolated pieces |
| **Readability** | Named functions explain what the code does |

## 2. Defining and Calling Functions

### Anatomy of a function definition

```python
def function_name():
    # indented block — the function body
    # runs when the function is called
```

- **`def`** — keyword that starts a function definition
- **`function_name`** — the name you choose (use `snake_case` per PEP 8)
- **`()`** — parentheses (parameters go inside, if any)
- **`:`** — colon to start the function body
- **indented block** — the code that runs when you call the function

### Calling a function

To **call** (execute) a function, write its name followed by parentheses:

````python
# Define: write the function logic
def greet_soc_team():
    print("Hello, SOC team!")
    print("Ready to monitor threats?")

# Call: execute the function
greet_soc_team()
````

````python
# Call the same function multiple times
greet_soc_team()
print()  # blank line for readability
greet_soc_team()
````

### ⚠️ Common mistake: forgetting the parentheses

```python
greet_soc_team    # This does NOT call the function — it just refers to it
greet_soc_team()  # This DOES call the function
```

## 3. Parameters — Inputs to Functions

### What are parameters?

**Parameters** (also called arguments) are **inputs** to a function — values you pass in when calling it.

```python
def function_name(parameter1, parameter2):
    # use parameter1 and parameter2 inside the function
```

### Function with one parameter

````python
# Function with one parameter: username
def greet_analyst(username):
    print(f"Hello, {username}!")
    print("Welcome to the Security Operations Center")

# Call with different usernames
greet_analyst("Alice")
print()
greet_analyst("Bob")
````

### Function with multiple parameters

````python
# Function with two parameters: IP and port
def check_service(ip_address, port_number):
    print(f"Checking {ip_address} on port {port_number}...")
    if port_number == 22:
        print("Service: SSH")
    elif port_number == 80:
        print("Service: HTTP")
    elif port_number == 443:
        print("Service: HTTPS")
    else:
        print("Service: Unknown")

# Call with different IPs and ports
check_service("192.168.1.10", 22)
print()
check_service("192.168.1.11", 443)
````

### Security example: password strength checker

````python
# Check if a password meets minimum length policy
def check_password_length(password, min_length):
    actual_length = len(password)
    print(f"Password length: {actual_length}")
    print(f"Minimum required: {min_length}")
    
    if actual_length >= min_length:
        print("✓ Password meets length requirement")
    else:
        gap = min_length - actual_length
        print(f"✗ Password too short by {gap} character(s)")

# Test with different passwords
check_password_length("abc123", 12)
print()
check_password_length("Secure!Pass#2024", 12)
````

## 4. Return Values — Outputs from Functions

### What is `return`?

The `return` statement **sends a value back** from the function to the caller. Think of it as the **output** of the function.

```python
def add_numbers(a, b):
    result = a + b
    return result  # send result back to the caller
```

### Using the returned value

````python
# Function that returns a value
def calculate_cvss_base(impact, exploitability):
    # Simplified CVSS calculation
    base_score = (impact + exploitability) / 2
    return base_score

# Capture the returned value
score = calculate_cvss_base(8, 9)
print(f"CVSS base score: {score}")

# Use the returned value in a comparison
if score >= 7:
    print("Severity: HIGH")
else:
    print("Severity: MEDIUM")
````

### Returning boolean values

````python
# Function that returns True or False
def is_high_port(port_number):
    return port_number >= 1024

# Use the boolean result in an if statement
port = 8080
if is_high_port(port):
    print(f"Port {port} is a high (unprivileged) port")
else:
    print(f"Port {port} is a well-known (privileged) port")
````

### Security example: validate CVE ID format

````python
# Check if a string matches the CVE ID format CVE-YYYY-NNNN
def is_valid_cve_id(cve_string):
    # Basic validation: starts with 'CVE-' and has 12+ characters
    if not cve_string.startswith("CVE-"):
        return False
    if len(cve_string) < 12:
        return False
    return True

# Test with different inputs
test_ids = ["CVE-2024-1234", "CVE-2024", "2024-1234", "CVE-2024-56789"]

for test_id in test_ids:
    if is_valid_cve_id(test_id):
        print(f"✓ {test_id} is valid")
    else:
        print(f"✗ {test_id} is invalid")
````

## 5. Default Parameters

### What are default parameters?

You can give a parameter a **default value**. If the caller does not provide that argument, the default is used.

```python
def function_name(param1, param2=default_value):
    ...
```

### Why use defaults?

- The caller can omit common values
- Makes functions easier to call in typical cases
- Keeps flexibility for special cases

````python
# Function with a default parameter: protocol defaults to "https"
def build_url(hostname, protocol="https"):
    return f"{protocol}://{hostname}"

# Call with both arguments
url1 = build_url("example.com", "http")
print(url1)  # http://example.com

# Call with only the required argument — uses default protocol
url2 = build_url("secure.example.com")
print(url2)  # https://secure.example.com
````

### Security example: scan with default timeout

````python
# Simulate a port scan with a configurable timeout (default 2 seconds)
def scan_port_with_timeout(ip_address, port_number, timeout_seconds=2):
    print(f"Scanning {ip_address}:{port_number} with {timeout_seconds}s timeout...")
    # Simulate result
    return "open"

# Use default timeout
result1 = scan_port_with_timeout("192.168.1.10", 80)
print(f"Result: {result1}\n")

# Override timeout for a slow network
result2 = scan_port_with_timeout("192.168.1.11", 443, timeout_seconds=5)
print(f"Result: {result2}")
````

## 6. `return` vs `print()`

### The critical difference

Beginners often confuse `return` and `print()`. They are **completely different**:

| Statement | What it does | Where the value goes |
|---|---|---|
| `print()` | Displays text on screen | The terminal / notebook output |
| `return` | Sends a value back to the caller | The variable that captures it |

### A function that prints but does not return

````python
# Function that prints — no return value
def print_scan_result(port_number):
    print(f"Port {port_number} is open")

# Call it and try to capture the result
result = print_scan_result(80)

print(f"\nCaptured result: {result}")  # None — the function returned nothing
````

### A function that returns but does not print

````python
# Function that returns — no print
def get_scan_result(port_number):
    return f"Port {port_number} is open"

# Call it and capture the result
result = get_scan_result(443)

# Nothing printed yet — the value is just stored in 'result'
print(f"Captured result: {result}")

# Now print the result
print(result)
````

### When to use each

- **Use `return`** when you need the function to **produce a value** that the caller will use (calculations, validations, transformations)
- **Use `print()`** when you want to **show information to the user** (logging, debugging, user-facing messages)
- You can use both in the same function if needed

## Summary

| Concept | Syntax | Purpose |
|---|---|---|
| Define a function | `def name():` | Create a reusable block of code |
| Call a function | `name()` | Execute the function |
| Parameter | `def name(param):` | Input to the function |
| Return value | `return value` | Output from the function |
| Default parameter | `def name(param=default):` | Optional input with fallback value |

### Key rules to remember

- Always indent the function body (4 spaces per PEP 8)
- Parameters are **inputs**, `return` sends **outputs**
- `print()` displays text; `return` sends a value back to the caller
- Default parameters must come **after** required parameters
- A function without `return` implicitly returns `None`
