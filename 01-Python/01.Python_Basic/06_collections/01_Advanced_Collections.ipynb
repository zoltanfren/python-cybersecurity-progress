## 1. Negative Indexing — Counting from the End

### How It Works

Positive indexes count from the **start** (left to right). Negative indexes count from the **end** (right to left).

```
ports = [ 80,  443, 8080,   22, 3306]

Positive:   0     1     2     3     4     (left to right)
Negative:  -5    -4    -3    -2    -1     (right to left)
```

`-1` is always the **last item**, regardless of list length.

**Why is this useful?** When you need the last item but you don't know the length of the list. Without negative indexing, you would need `my_list[len(my_list) - 1]`. With it, just `my_list[-1]`.

````python
ports = [80, 443, 8080, 22, 3306]
#          0    1     2   3     4   (positive)
#         -5   -4    -3  -2    -1   (negative)

print("ports[-1]  (last):",        ports[-1])   # 3306
print("ports[-2]  (second last):", ports[-2])   # 22
print("ports[-5]  (first):",       ports[-5])   # 80  — same as ports[0]
print()

# Useful: get last item without knowing the length
log_entries = ["INFO: scan started", "INFO: port 80 open", "WARN: port 23 open"]
print("Latest log entry:", log_entries[-1])
````

````python
# Works on strings too — strings are sequences of characters
ip = "192.168.1.100"

print("Last character:",  ip[-1])   # 0
print("Second to last:",  ip[-2])   # 0

# Works on tuples too
location = ("Brussels", "Belgium", "Europe")
print("Last item:", location[-1])   # Europe
````

## 2. Slicing — Extracting Subsets

### What Is Slicing?

Slicing lets you extract a **portion** of a list (or string, or tuple) without modifying the original.

**Syntax:**
```python
my_list[start : stop : step]
```

| Part | Meaning | Default |
|------|---------|--------|
| `start` | Index to start from (inclusive) | `0` (beginning) |
| `stop` | Index to stop at (**exclusive** — this item is NOT included) | `len(list)` (end) |
| `step` | How many positions to jump each time | `1` |

### The Golden Rule

`stop` is **exclusive** — the item at that index is **not** included in the result.

```
ports = [80, 443, 8080, 22, 3306]
           0    1     2   3     4

ports[1:3]  →  [443, 8080]    (indexes 1 and 2 — NOT 3)
```

````python
ports = [80, 443, 8080, 22, 3306]
#          0    1     2   3     4

# Basic slices
print("ports[1:3]  → indexes 1,2:    ", ports[1:3])    # [443, 8080]
print("ports[0:2]  → indexes 0,1:    ", ports[0:2])    # [80, 443]
print("ports[2:5]  → indexes 2,3,4:  ", ports[2:5])    # [8080, 22, 3306]
````

### Omitting start or stop

When you omit `start`, Python uses `0`. When you omit `stop`, Python uses the end of the list.

````python
ports = [80, 443, 8080, 22, 3306]

print("ports[:3]   → first 3 items:  ", ports[:3])     # [80, 443, 8080]
print("ports[2:]   → from index 2:   ", ports[2:])     # [8080, 22, 3306]
print("ports[:]    → full copy:      ", ports[:])      # [80, 443, 8080, 22, 3306]
````

### Combining Negative Indexes with Slicing

````python
ports = [80, 443, 8080, 22, 3306]

print("ports[-3:]   → last 3:        ", ports[-3:])    # [8080, 22, 3306]
print("ports[:-2]   → all but last 2:", ports[:-2])    # [80, 443, 8080]
print("ports[-3:-1] → 3rd & 2nd last:", ports[-3:-1])  # [8080, 22]
````

### The Step Parameter

The `step` controls how many positions to jump between each selected item.

```
my_list[::2]   → every 2nd item (skip 1 between each)
my_list[::-1]  → step of -1 → traverse backwards → reverses the list
```

````python
ports = [80, 443, 8080, 22, 3306, 21, 23]

print("ports[::2]   → every 2nd:  ", ports[::2])    # [80, 8080, 3306, 23]
print("ports[::3]   → every 3rd:  ", ports[::3])    # [80, 22, 23]
print("ports[::-1]  → reversed:   ", ports[::-1])   # [23, 21, 3306, 22, 8080, 443, 80]
````

````python
# Slicing works on strings too — very useful for parsing
log_line = "2024-01-15 10:23:45 ERROR Failed login from 192.168.1.100"

date      = log_line[:10]   # first 10 chars
time_part = log_line[11:19] # chars 11 to 18
severity  = log_line[20:25] # chars 20 to 24

print("Date:    ", date)
print("Time:    ", time_part)
print("Severity:", severity)
````

### Slicing Does NOT Modify the Original

A slice always returns a **new list**. The original is unchanged.

````python
ports = [80, 443, 8080, 22, 3306]

first_three = ports[:3]   # creates a new list
print("Slice:    ", first_three)  # [80, 443, 8080]
print("Original: ", ports)        # [80, 443, 8080, 22, 3306] — unchanged
````

## 3. List Methods in Depth

### Sorting Lists

````python
ports = [443, 80, 22, 3306, 8080, 21]
print("Original:", ports)

# sort() — modifies the list IN PLACE
ports.sort()
print("Sorted ascending: ", ports)

ports.sort(reverse=True)
print("Sorted descending:", ports)

# sorted() — returns a NEW sorted list, original unchanged
original = [443, 80, 22, 3306]
new_sorted = sorted(original)
print("sorted() result:",   new_sorted)
print("Original intact:  ", original)
````

### Copying Lists — The Mutation Trap

This is one of the most common beginner mistakes. When you do `b = a`, you are NOT copying the list — both variables point to the **same object in memory**.

````python
# THE TRAP: assignment does NOT copy
original = [22, 80, 443]
alias    = original          # both point to the same list!

alias.append(3306)           # modifying alias also modifies original!
print("Alias:   ", alias)
print("Original:", original) # also changed!
print()

# THE FIX: create a real copy
original = [22, 80, 443]
copy1 = original.copy()      # Method 1: .copy()
copy2 = original[:]          # Method 2: slice — creates new list
copy3 = list(original)       # Method 3: list() constructor

copy1.append(9999)
print("copy1 modified:", copy1)
print("original safe: ", original)  # unchanged!
````

### Combining Lists

````python
web_ports = [80, 443, 8080]
db_ports  = [3306, 5432, 1521]

# + operator — creates a new combined list
all_ports = web_ports + db_ports
print("Combined with +:      ", all_ports)

# extend() — adds all items from another list IN PLACE
web_ports.extend(db_ports)
print("Extended (in place):  ", web_ports)
````

## 4. Set Operations — Comparing Collections

### Why Sets Are Powerful

Sets support **mathematical set operations**: comparing two groups of items to find what they share, what's unique, or what's in one but not the other. These are extremely useful in security work.

| Operation | Symbol | Method | Meaning |
|-----------|--------|--------|---------|
| Union | `\|` | `.union()` | Everything in either set |
| Intersection | `&` | `.intersection()` | Only what is in BOTH |
| Difference | `-` | `.difference()` | In A but NOT in B |
| Symmetric Difference | `^` | `.symmetric_difference()` | In one but NOT both |

````python
# Compare open ports across two servers
server1_ports = {22, 80, 443, 3306}
server2_ports = {22, 80, 443, 5432}

print("Server 1:", server1_ports)
print("Server 2:", server2_ports)
print()

# Union — all ports from both
all_ports = server1_ports | server2_ports
print("Union (all ports):         ", all_ports)

# Intersection — ports open on BOTH servers
shared_ports = server1_ports & server2_ports
print("Intersection (shared):     ", shared_ports)

# Difference — open on server1 but NOT server2
only_server1 = server1_ports - server2_ports
print("Difference (only in s1):   ", only_server1)

# Symmetric difference — open on one server but not both
not_shared = server1_ports ^ server2_ports
print("Sym. difference (not both):", not_shared)
````

````python
# Real security use: find IPs that appeared in yesterday's AND today's logs
# These could be persistent attackers worth investigating

yesterday_ips = {"10.0.0.50", "192.168.1.100", "8.8.8.8", "1.2.3.4"}
today_ips     = {"10.0.0.50", "192.168.1.100", "5.6.7.8", "9.10.11.12"}

# IPs in both — potential persistent threat
persistent = yesterday_ips & today_ips
print("Persistent IPs (in both days):", persistent)

# New IPs today that were not seen yesterday
new_today = today_ips - yesterday_ips
print("New today (not seen yesterday):", new_today)

# Total unique IPs over both days
total_unique = yesterday_ips | today_ips
print("Total unique IPs:              ", total_unique)
````

### Set Membership — Why Sets Beat Lists for Lookups

Checking `x in my_list` scans every item until found — slow for large lists (O(n)).
Checking `x in my_set` uses a hash table — instant regardless of size (O(1)).

**Rule of thumb:** If you have a large collection and need to check membership frequently, use a `set` or `dict` instead of a `list`.

````python
# Build a blocklist of 10,000 IPs
blocked_as_list = [f"10.0.{i//256}.{i%256}" for i in range(10000)]
blocked_as_set  = set(blocked_as_list)

test_ip = "10.0.39.16"

# Both give the correct answer — but set is massively faster at scale
print("In list:", test_ip in blocked_as_list)  # scans up to 10,000 items
print("In set: ", test_ip in blocked_as_set)   # instant hash lookup
print()
print("For a blocklist of 10,000 IPs:")
print("  List: checks up to 10,000 items one by one")
print("  Set:  answers instantly using a hash table")
````

## 5. List Comprehensions — Concise Filtering and Transformation

### What Is a List Comprehension?

A list comprehension builds a new list from an existing one in a single, readable line. It is the Pythonic alternative to a `for` loop that appends to a list.

**Syntax:**
```python
[expression for item in iterable if condition]
```

Reading it: "Give me `expression` for each `item` in `iterable`, but only when `condition` is True."

The `if condition` part is **optional**.

````python
# Traditional: filter with a for loop
all_ports = [21, 22, 80, 443, 3306, 8080]

safe_ports_loop = []
for p in all_ports:
    if p > 1024:
        safe_ports_loop.append(p)
print("Loop result:          ", safe_ports_loop)

# Same thing with a list comprehension
safe_ports_comp = [p for p in all_ports if p > 1024]
print("Comprehension result: ", safe_ports_comp)
````

````python
# Comprehension WITH transformation — change each item
usernames = ["Alice", "BOB", "Charlie"]

# Normalise all usernames to lowercase
normalised = [u.lower() for u in usernames]
print("Normalised:", normalised)

# Build formatted strings
ports = [22, 80, 443]
messages = [f"Port {p} is open" for p in ports]
for msg in messages:
    print(msg)
````

````python
# Filter AND transform at the same time
open_ports   = [21, 22, 80, 443, 23, 3306]
dangerous    = [21, 23, 135, 445]

# Get labels for only the dangerous ports that are open
alerts = [f"ALERT: port {p}" for p in open_ports if p in dangerous]
for alert in alerts:
    print(alert)
````

### When NOT to Use List Comprehensions

List comprehensions are powerful but should stay readable. If the logic is complex, use a regular loop.

```python
# Too complex for a comprehension — use a loop
result = [complicated_function(x) for x in data if some_condition(x) and another(x)]

# Better as a loop when logic is multi-step
result = []
for x in data:
    if some_condition(x) and another(x):
        result.append(complicated_function(x))
```

**Rule:** If a comprehension needs a second read to understand, write a loop instead.

## 6. Nested Collections — Real-World Data Structures

Real security data is never flat. You often need collections inside collections.

### List of Dictionaries — The Table Pattern

Think of it as a database table: each dictionary is a row, each key is a column.

````python
# Vulnerability scan results — list of findings
vulnerabilities = [
    {"cve": "CVE-2023-1234", "cvss": 9.8, "component": "OpenSSL",   "patched": False},
    {"cve": "CVE-2023-5678", "cvss": 7.2, "component": "Apache",    "patched": True},
    {"cve": "CVE-2023-9012", "cvss": 5.0, "component": "nginx",     "patched": False},
    {"cve": "CVE-2023-3456", "cvss": 9.1, "component": "PHP",       "patched": False},
]

# Access the first finding
print("First finding:", vulnerabilities[0])
print()

# Access a specific field
print("First CVE ID:",  vulnerabilities[0]["cve"])
print("First CVSS:  ",  vulnerabilities[0]["cvss"])
````

````python
# List comprehension on nested structures

# Get only unpatched critical vulnerabilities (cvss >= 9.0)
critical_unpatched = [
    v for v in vulnerabilities
    if v["cvss"] >= 9.0 and not v["patched"]
]

print("Critical unpatched vulnerabilities:")
for v in critical_unpatched:
    print(f"  {v['cve']} — CVSS {v['cvss']} — {v['component']}")
````

### Nested Dictionary — The Config Pattern

````python
# Server infrastructure config
infrastructure = {
    "web01": {
        "ip":       "192.168.1.100",
        "ports":    [80, 443],
        "services": ["nginx", "php-fpm"],
        "patched":  True
    },
    "db01": {
        "ip":       "192.168.1.101",
        "ports":    [3306],
        "services": ["mysql"],
        "patched":  False
    }
}

# Access nested data
print("web01 IP:",       infrastructure["web01"]["ip"])
print("web01 ports:",    infrastructure["web01"]["ports"])
print("db01 services:",  infrastructure["db01"]["services"])
print()

# Find unpatched servers
unpatched = [name for name, cfg in infrastructure.items() if not cfg["patched"]]
print("Unpatched servers:", unpatched)
````

## 📝 Summary

| Concept | Syntax | Key rule |
|---------|--------|----------|
| **Negative indexing** | `my_list[-1]` | `-1` = last, `-2` = second last |
| **Slicing** | `my_list[start:stop:step]` | `stop` is exclusive — item at `stop` is NOT included |
| **Omitting start/stop** | `[:3]`, `[2:]`, `[:]` | Defaults to 0 and end of list |
| **Reverse** | `[::-1]` | Step of -1 traverses backwards |
| **Slice = new list** | `copy = original[:]` | Original is never modified by slicing |
| **copy() vs alias** | `b = a.copy()` vs `b = a` | `b = a` does NOT copy — both point to same object |
| **Set union** | `A \| B` | All items from both sets |
| **Set intersection** | `A & B` | Only items in both |
| **Set difference** | `A - B` | In A but not in B |
| **List comprehension** | `[expr for x in l if cond]` | Build a new list in one readable line |
| **Set for lookups** | `x in my_set` | O(1) — instant, regardless of size |
