# Module 7 - Collections Beginner Level

In this notebook, you will learn how to store and organize multiple pieces of data using Python's four collection types.

### What you will learn:
- Why collections exist and what problem they solve
- **Lists** — ordered, changeable collections
- **Tuples** — ordered, fixed collections
- **Dictionaries** — key-value lookups
- **Sets** — unique values only
- When to use each one

---
## Why Collections?

Imagine you are managing a server and need to track open ports:

```python
# Without collections — gets messy fast
port1 = 80
port2 = 443
port3 = 8080
port4 = 22
port5 = 3306
# What if you have 50 ports?
```

**With a collection:**
```python
ports = [80, 443, 8080, 22, 3306]
```

One variable, all the data. Collections let you group related values together and work with them as a unit.

### The Four Collection Types

| Type | Syntax | Ordered? | Changeable? | Duplicates? | Best for |
|------|--------|----------|-------------|-------------|----------|
| **List** | `[a, b, c]` | Yes | Yes | Yes | Dynamic sequences |
| **Tuple** | `(a, b, c)` | Yes | No | Yes | Fixed data |
| **Dictionary** | `{k: v}` | Yes* | Yes | Keys: No | Key-value lookups |
| **Set** | `{a, b, c}` | No | Yes | No | Unique values |

*Dictionary order is preserved in Python 3.7+

## 1. Lists

### What Is a List?

A **list** is an ordered sequence of items. Think of it as a numbered shelf:
- The order is preserved
- You can add, remove, or change items
- The same value can appear more than once
- Every item has a **position number** called an index, starting at `0`

### Creating a List

Lists are written with square brackets `[]`, items separated by commas:

````python
# Lists can hold any type of data
open_ports  = [80, 443, 8080, 22]
usernames   = ["alice", "bob", "charlie"]
mixed       = ["web01", 80, True]     # different types in one list
empty_list  = []                       # empty list — you can fill it later

print("Ports:",    open_ports)
print("Users:",    usernames)
print("Mixed:",    mixed)
print("Empty:",    empty_list)
````

### Accessing Items — Indexing

Every item in a list has an index. **Python always starts counting at 0.**

```
ports = [ 80,  443, 8080,   22, 3306]
           |     |     |     |     |
Index:     0     1     2     3     4
```

To get item at index `i`, write `my_list[i]`.

````python
ports = [80, 443, 8080, 22, 3306]

print("Index 0 (first):",  ports[0])   # 80
print("Index 1 (second):", ports[1])   # 443
print("Index 4 (last):",   ports[4])   # 3306
print("Length:",           len(ports)) # 5
````

````python
# ⚠️  Accessing an index that doesn't exist causes an IndexError
# ports[10]  # IndexError: list index out of range

# Safe pattern: check length before accessing
ports = [80, 443, 22]
index_to_check = 10

if index_to_check < len(ports):
    print(ports[index_to_check])
else:
    print(f"No item at index {index_to_check} — list has only {len(ports)} items")
````

### Adding Items

````python
ports = [80, 443]
print("Before:", ports)

# append() — adds to the END of the list
ports.append(22)
print("After append(22):", ports)

# insert(index, value) — adds at a SPECIFIC position
ports.insert(0, 21)   # insert 21 at index 0 (beginning)
print("After insert(0, 21):", ports)
````

### Removing Items

````python
ports = [80, 443, 8080, 22]
print("Before:", ports)

# remove(value) — removes first occurrence of a VALUE
ports.remove(8080)
print("After remove(8080):", ports)

# pop(index) — removes item at an INDEX and returns it
removed = ports.pop(0)   # remove item at index 0
print(f"Removed: {removed}")
print("After pop(0):", ports)
````

### Checking Membership

````python
open_ports    = [22, 80, 443, 8080]
dangerous_ports = [21, 23, 135, 445, 3389]

# Use 'in' to check if a value exists
print("Is 443 open?",  443 in open_ports)      # True
print("Is 9000 open?", 9000 in open_ports)     # False
print("Is 23 dangerous?", 23 in dangerous_ports) # True

# Use 'not in' to check absence
if 21 not in open_ports:
    print("Good — FTP port 21 is closed")
````

## 2. Tuples

### What Is a Tuple?

A **tuple** looks like a list, but it is **immutable** — once created, you cannot change it.

You cannot add, remove, or modify items in a tuple.

### Creating a Tuple

Tuples use parentheses `()` instead of square brackets:

````python
# Tuples — fixed data
server_location = ("Brussels", "Belgium")
coordinates     = (50.8503, 4.3517)
http_ports      = (80, 443)        # these standard ports don't change

print("Location:",    server_location)
print("Coordinates:", coordinates)
print("HTTP ports:",  http_ports)
````

````python
# Accessing tuple items — same as lists, use index
location = ("Brussels", "Belgium", "Europe")

print("City:",      location[0])   # Brussels
print("Country:",   location[1])   # Belgium
print("Continent:", location[2])   # Europe
````

````python
# Tuples cannot be changed — uncomment to see the error
ports = (80, 443)

ports.append(8080)   # AttributeError: 'tuple' object has no attribute 'append'
ports[0] = 90        # TypeError: 'tuple' object does not support item assignment

print("Tuple is safe from accidental modification:", ports)
````

### Tuple Unpacking

A useful feature: you can assign tuple values to multiple variables at once:

````python
# Tuple unpacking — assign each value to a variable
server_info = ("web01", "192.168.1.100", 80)

name, ip, port = server_info   # unpack all three at once

print(f"Server: {name}")
print(f"IP:     {ip}")
print(f"Port:   {port}")
````

## 3. Dictionaries

### What Is a Dictionary?

A **dictionary** stores data as **key-value pairs** — like a real dictionary where a word (key) maps to its definition (value).

Instead of accessing values by position (like lists), you access them **by key**.

**Real-world analogy:** A phone book.
- Key = person's name
- Value = their phone number

You look up the name to get the number — not the position.

### Creating a Dictionary

Dictionaries use curly braces `{}` with `key: value` pairs:

````python
# Port number → service name mapping
port_services = {
    22:   "SSH",
    80:   "HTTP",
    443:  "HTTPS",
    3306: "MySQL"
}

print(port_services)
````

````python
# Keys can be strings too
user = {
    "username": "alice",
    "role":     "admin",
    "active":   True,
    "logins":   42
}

print(user)
````

### Accessing Values

````python
port_services = {22: "SSH", 80: "HTTP", 443: "HTTPS"}

# Access by key — use square brackets with the KEY (not an index)
print("Port 22:",  port_services[22])    # SSH
print("Port 443:", port_services[443])   # HTTPS

# ⚠️  Accessing a missing key causes a KeyError
# port_services[9999]   # KeyError: 9999

# Safe alternative: .get() returns None (or a default) if key is missing
service = port_services.get(9999, "Unknown")
print("Port 9999:", service)   # Unknown
````

### Adding, Updating, and Removing

````python
services = {
    22: "SSH", 
    80: "HTTP"
}
print("Start:", services)

# Add new key-value pair
services[443] = "HTTPS"
print("Added 443:", services)

# Update existing value
services[80] = "HTTP (legacy)"
print("Updated 80:", services)

# Remove a key
del services[22]
print("Removed 22:", services)
````

### Useful Dictionary Methods

````python
port_services = {
    22: "SSH", 
    80: "HTTP", 
    443: "HTTPS"
}

# Check if a key exists
print("80 in dict?:", 80 in port_services)    # True
print("99 in dict?:", 99 in port_services)    # False

# Get all keys, values, or pairs
print("Keys:",   list(port_services.keys()))    # [22, 80, 443]
print("Values:", list(port_services.values()))  # ['SSH', 'HTTP', 'HTTPS']
print("Items:",  list(port_services.items()))   # [(22, 'SSH'), ...]
````

## 4. Sets

### What Is a Set?

A **set** is a collection of **unique values** with no duplicates and no guaranteed order.

**Real-world analogy:** A bag of unique stamps — you can have each stamp only once, and the order they are stored in doesn't matter.

Key properties:
- **No duplicates** — adding the same value twice has no effect
- **No index** — you cannot do `my_set[0]`
- **Fast membership testing** — checking `x in my_set` is very fast

### Creating a Set

````python
# Sets use curly braces {}
unique_ports = {80, 443, 22, 8080}
print("Set:", unique_ports)   # order may vary!

# Duplicates are automatically removed
with_dupes = {80, 443, 80, 22, 443}
print("Duplicates removed:", with_dupes)
````

````python
# Convert a list to a set to remove duplicates
ip_addresses = ["192.168.1.1", "10.0.0.1", "192.168.1.1", "10.0.0.2", "10.0.0.1"]
print("List:",         ip_addresses)
print("Unique IPs:",   set(ip_addresses))
print("Duplicate count:", len(ip_addresses) - len(set(ip_addresses)))
````

### Adding to and Removing from a Set

````python
blocked_ips = {"10.0.0.50", "192.168.1.100"}
print("Before:", blocked_ips)

# add() — adds a value (no effect if already in set)
blocked_ips.add("172.16.0.25")
print("After add:", blocked_ips)

blocked_ips.add("10.0.0.50")   # already exists — no change
print("After duplicate add:", blocked_ips)

# discard() — removes a value (no error if not found)
blocked_ips.discard("10.0.0.50")
print("After discard:", blocked_ips)
````

## 5. When to Use Which Collection?

This is the most important question in practice. The right choice affects clarity, safety, and performance.

### Decision Guide

Ask yourself these questions:

```
Do I need key-value lookup?
  YES → Dictionary
  NO  → Continue...

Do I need uniqueness (no duplicates)?
  YES → Set
  NO  → Continue...

Will the data change (add/remove items)?
  YES → List
  NO  → Tuple
```

### Concrete Examples

| Situation | Best choice | Why |
|-----------|-------------|-----|
| Ports discovered during a scan | `list` | Ordered, will grow as you find more, duplicates possible |
| Port number → service name | `dict` | You look up by port number (key) |
| Server GPS coordinates | `tuple` | Fixed, never changes, two values (lat, lon) |
| List of unique blocked IPs | `set` | No duplicates, fast `in` checks |
| Ordered log entries | `list` | Order matters (chronological), grows over time |
| Username → permissions | `dict` | Look up permissions by username |
| Standard HTTP ports `(80, 443)` | `tuple` | These ports don't change — protect them |
| Unique users who logged in today | `set` | Unique only, no order needed |

````python
# Example: choosing the right collection for each security task

# Scan results grow as we find ports — LIST
open_ports = []
open_ports.append(22)
open_ports.append(80)
open_ports.append(443)

# Look up service by port number — DICTIONARY
port_services = {22: "SSH", 80: "HTTP", 443: "HTTPS"}

# Server location never changes — TUPLE
data_center = ("Brussels", "Belgium", 50.8503, 4.3517)

# Track unique IPs we have seen — SET
seen_ips = set()
seen_ips.add("192.168.1.1")
seen_ips.add("10.0.0.1")
seen_ips.add("192.168.1.1")  # no duplicate added

print("Open ports (list):",   open_ports)
print("Services (dict):",     port_services)
print("Location (tuple):",    data_center)
print("Seen IPs (set):",      seen_ips)
````

## 6. Nested Collections

Collections can contain other collections. This is how real-world data is often structured.

### List of Dictionaries — The Most Common Pattern

````python
# Each item in the list is a dictionary — like a database table
servers = [
    {
        "name": "web01", 
        "ip": "192.168.1.100", 
        "port": 80
    },
    {
        "name": "web02", 
        "ip": "192.168.1.101", 
        "port": 80
    },
    {
        "name": "db01",  
        "ip": "192.168.1.102", 
        "port": 3306
    }
]

# Access the first server
print("First server:", servers[0])

# Access a specific field of the first server
print("First server IP:", servers[0]["ip"])
````

### Dictionary with List Values

````python
# User permissions — each user has a list of permissions
user_permissions = {
    "alice":   ["read", "write", "execute"],
    "bob":     ["read"],
    "charlie": ["read", "write"]
}

# Check if alice can write
if "write" in user_permissions["alice"]:
    print("Alice can write")

# Add a permission to bob
user_permissions["bob"].append("write")
print("Bob's permissions:", user_permissions["bob"])
````

## 📝 Summary

| Collection | Syntax | Key features | Use when... |
|------------|--------|--------------|-------------|
| **List** | `[a, b]` | Ordered, changeable, allows duplicates | You need a sequence that grows or changes |
| **Tuple** | `(a, b)` | Ordered, fixed, allows duplicates | Data must stay constant (coordinates, config) |
| **Dictionary** | `{k: v}` | Key-value pairs, fast lookup by key | You need to look things up by name or ID |
| **Set** | `{a, b}` | No duplicates, no order, fast `in` check | You need unique values or fast membership tests |
