````python
import unittest


class TestCollections(unittest.TestCase):

    # --- Part 1: Lists ---
    def test_01_target_servers(self):
        self.assertEqual(
            target_servers, ["web01", "web02", "db01"],
            "target_servers should be exactly ['web01', 'web02', 'db01'] in that order"
        )

    def test_02_second_server(self):
        self.assertEqual(
            second_server, "web02",
            "second_server should be 'web02' — Python indices start at 0, so the second item is index 1"
        )

    def test_03_slicing(self):
        self.assertEqual(
            first_three_ports, [21, 22, 80],
            "first_three_ports should be [21, 22, 80] — use all_ports[:3]"
        )
        self.assertEqual(
            last_port, 3306,
            "last_port should be 3306 — use all_ports[-1]"
        )

    # --- Part 2: Tuples and Sets ---
    def test_04_server_location(self):
        self.assertEqual(
            server_location, ("Brussels", "Belgium"),
            "server_location should be ('Brussels', 'Belgium')"
        )
        self.assertIsInstance(
            server_location, tuple,
            "server_location must be a TUPLE — use () not []"
        )

    def test_05_unique_ips(self):
        self.assertEqual(
            unique_ips, {"192.168.1.1", "192.168.1.2", "10.0.0.1"},
            "unique_ips should have 3 elements — '192.168.1.1' appeared twice but sets keep only one copy"
        )
        self.assertIsInstance(
            unique_ips, set,
            "unique_ips must be a SET — use set(ip_addresses)"
        )

    def test_06_common_ports(self):
        self.assertEqual(
            common_ports, {22, 80, 443},
            "common_ports should be {22, 80, 443} — use the intersection operator: server_a_ports & server_b_ports"
        )

    # --- Part 3: Dictionaries ---
    def test_07_port_services(self):
        self.assertEqual(port_services[22],  "SSH",   "port_services[22] should be 'SSH'")
        self.assertEqual(port_services[80],  "HTTP",  "port_services[80] should be 'HTTP'")
        self.assertEqual(port_services[443], "HTTPS", "port_services[443] should be 'HTTPS'")

    def test_08_port_services_mysql(self):
        self.assertEqual(
            port_services[3306], "MySQL",
            "port_services[3306] should be 'MySQL' — use port_services[3306] = 'MySQL'"
        )

    def test_09_server_config_and_web_ip(self):
        self.assertEqual(server_config["web01"]["ip"],   "192.168.1.100", "server_config['web01']['ip'] should be '192.168.1.100'")
        self.assertEqual(server_config["web01"]["port"],  80,             "server_config['web01']['port'] should be 80 (integer)")
        self.assertEqual(server_config["db01"]["ip"],    "192.168.1.101", "server_config['db01']['ip'] should be '192.168.1.101'")
        self.assertEqual(server_config["db01"]["port"],   3306,           "server_config['db01']['port'] should be 3306 (integer)")
        self.assertEqual(
            web_ip, "192.168.1.100",
            "web_ip should be '192.168.1.100' — access with server_config['web01']['ip']"
        )

    def test_10_user_permissions(self):
        self.assertEqual(user_permissions["alice"],   ["read", "write"],            "alice should have ['read', 'write']")
        self.assertEqual(user_permissions["bob"],     ["read"],                     "bob should have ['read'] only")
        self.assertEqual(user_permissions["charlie"], ["read", "write", "execute"], "charlie should have ['read', 'write', 'execute']")

    # --- Part 4: Comprehensions ---
    def test_11_high_ports(self):
        self.assertEqual(
            high_ports, [3306],
            "high_ports should be [3306] — only 3306 is >= 1024 in [21, 22, 80, 443, 3306]"
        )

    def test_12_failed_lines(self):
        self.assertEqual(
            len(failed_lines), 2,
            "failed_lines should have exactly 2 entries (the two lines that contain 'FAILED')"
        )
        self.assertTrue(
            all("FAILED" in line for line in failed_lines),
            "Every entry in failed_lines must contain the word 'FAILED'"
        )

    # --- Part 5 ---
    def test_13_slicing_operations(self):
        self.assertEqual(
            recent_alerts, ["alert_D", "alert_E"],
            "recent_alerts should be ['alert_D', 'alert_E'] — use alerts[-2:]"
        )
        self.assertEqual(
            oldest_alert, "alert_A",
            "oldest_alert should be 'alert_A' — use alerts[0]"
        )
        self.assertEqual(
            every_other, ["alert_A", "alert_C", "alert_E"],
            "every_other should be ['alert_A', 'alert_C', 'alert_E'] — use alerts[::2]"
        )

    def test_14_safe_lookup(self):
        self.assertEqual(
            unknown_service, "Unknown",
            "unknown_service should be 'Unknown' — use port_services_lookup.get(9999, 'Unknown')"
        )

    def test_15_safe_and_risky_ports(self):
        self.assertEqual(
            sorted(safe_ports), [22, 80, 443, 8080],
            "safe_ports should be [22, 80, 443, 8080] — ports NOT in the dangerous set"
        )
        self.assertEqual(
            sorted(risky_ports), [21, 135],
            "risky_ports should be [21, 135] — ports from open_ports that ARE in dangerous"
        )


unittest.main(argv=[""], verbosity=2, exit=False)
````
